````python
# ============================================================
# SECTION 00 — Environment Setup
# Maximal Institutional Research Pipeline Notebook
# ============================================================

import os
import sys
import logging
import warnings
from pathlib import Path


import numpy as np
import pandas as pd
import xgboost as xgb
import lightgbm as lgb

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
 
# ML Models
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, classification_report
)

# Set up notebook display
pd.set_option("display.max_rows", 200)
pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 200)
pd.set_option("display.float_format", lambda x: f"{x:,.6f}")

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (14, 6)

warnings.filterwarnings("ignore")

# ============================================================
# Logging
# ============================================================  
logger = logging.getLogger("RGNB")
logger.setLevel(logging.INFO)
# Prevent duplicate handlers if you re-run the cell
if not logger.handlers:
    handler = logging.StreamHandler(sys.stdout)
    formatter = logging.Formatter("[%(asctime)s] %(levelname)s — %(message)s", "%H:%M:%S")
    handler.setFormatter(formatter)
    logger.addHandler(handler)
logger.propagate = False
logger.info("Environment intialized.")

# ============================================================
# Reproducibility
# ============================================================

SEED = 42
np.random.seed(SEED)


# ============================================================
# Project Structure Setup
# ============================================================

ROOT = Path(os.environ.get("MYQUANTLAB_ROOT", Path.cwd()))
if not (ROOT / "src").exists() and (ROOT.parent / "src").exists():
    ROOT = ROOT.parent   # demo: launched from notebooks/, step up to repo root
SRC_DIR =  ROOT / "src"
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))
DATA_DIR = ROOT / "data"
FEATURE_SAVE_DIR = ROOT / "artifacts" / "features"
LABEL_SAVE_DIR = ROOT / "artifacts" / "labels"
MODEL_SAVE_DIR = ROOT / "artifacts" / "models"
BACKTEST_SAVE_DIR = ROOT / "artifacts" / "backtests"
VISUAL_SAVE_DIR = ROOT / "artifacts" / "visuals"

# create directories if needed
for d in [
    DATA_DIR,
    FEATURE_SAVE_DIR,
    LABEL_SAVE_DIR,
    MODEL_SAVE_DIR,
    BACKTEST_SAVE_DIR,
    VISUAL_SAVE_DIR,
]:
    d.mkdir(parents=True, exist_ok=True)

logger.info("Directory structure validated.")


# ============================================================
# Import Your Maximal Modules (Features, Labels, CV, etc.)
# ============================================================

# Feature modules
import features.pipeline as feat_pipe

# Label modules
import labels.triple_barrier as lbl_tb
import labels.forward_returns as lbl_fwd
import labels.multi_horizon as lbl_mh
import labels.volatility_adj as lbl_voladj
# Add immediately after:
import labels.labels as lbl_main
import labels.meta_labeling as lbl_meta
import labels.barriers as lbl_barriers
import labels.events as lbl_events
import labels.volatility as lbl_vol

# - 2nd edition Label modules 
import labels.labels as lbl_main           # Core labeling functions
import labels.meta_labeling as lbl_meta    # Meta-labeling
import labels.barriers as lbl_barriers     # Barrier mechanics
import labels.events as lbl_events         # Event generation
import labels.volatility as lbl_vol        # Volatility computations


# Purged CV modules
import cv.cv_purged as cv_purged
import cv.cv_adaptive_embargo as cv_adapt
import cv.cv_combinatorial as cv_cpcv
import cv.cv_clustered as cv_cluster
import cv.cv_walkforward as cv_walk
import cv.cv_visual as cv_visual
import cv.cv_auditor as cv_auditor

# Signal Extensions
from extensions.signal_extensions import (
    quick_reversion,
    quick_conviction,
    quick_divergence,
    quick_sizing,
    quick_filter,
    run_all_extensions
)

# custom modules
import shap
from features.feature_descriptions import (
    get_top_features_df,
    print_section12_guide,
)
import plotly.graph_objects as go
from ipywidgets import interact, Dropdown, VBox, HBox, Output
import ipywidgets as widgets

# ============================================================
# Data Validation & Checking Utilities
# ============================================================
from utils import (
    # Core validation
    DataValidator,
    quick_check,
    
    # Dynamic column detection
    get_prob_columns,
    get_bull_col,
    get_bear_col,
    get_classes,
    safe_get_col,
    
    # Cross-sample analysis
    compare_across_samples,
    pattern_stability_check,
    validate_by_regime,
    
    # Run history tracking
    run_history_comparison,
    compare_with_saved,
)

# Initialize validator
validator = DataValidator()
logger.info("Data validation utilities loaded.")


from visualization.interaction_analysis import (
    indian_run_analysis,
    analyze_feature_pair
)
def get_feature_cols(df, exclude_patterns=None, min_valid_pct=0.5):
    """
    Dynamically extract feature columns from a DataFrame.
    
    Parameters:
    -----------
    df : DataFrame with features
    exclude_patterns : list of patterns to exclude (e.g., ['target', 'returns', 'true'])
    min_valid_pct : minimum percentage of non-NaN values required (0.5 = 50%)
    
    Returns:
    --------
    List of valid feature column names
    """
    if exclude_patterns is None:
        exclude_patterns = ['target', 'true', 'pred', 'signal', 'pnl', 'position', 
                           'exec', 'cost', 'trade', 'returns', 'price', 'Open', 
                           'High', 'Low', 'Close', 'Volume', 'prob_class']
    
    # Start with all columns
    cols = df.columns.tolist()
    
    # Exclude pattern matches
    feature_cols = []
    for col in cols:
        exclude = False
        for pattern in exclude_patterns:
            if pattern.lower() in col.lower():
                exclude = True
                break
        if not exclude:
            feature_cols.append(col)
    
    # Filter by valid percentage (handles NaNs from short datasets)
    n_rows = len(df)
    valid_cols = []
    for col in feature_cols:
        valid_pct = df[col].notna().sum() / n_rows
        if valid_pct >= min_valid_pct:
            valid_cols.append(col)
    
    return valid_cols

logger.info("All custom modules successfully imported.")


# ============================================================
# Done — Section 00
# ============================================================
````

````python
# ============================================================
# SECTION 00.1 — TIMEFRAME CONFIGURATION (Universal Scaling)
# ============================================================

from dataclasses import dataclass
from typing import List, Dict, Optional

@dataclass
class TimeframeConfig:
    """Universal timeframe configuration with auto-scaled parameters."""
    
    # --- Timeframe Definition ---
    timeframe: str              # e.g., "1H", "5M", "1D", "100T"
    bars_per_day: float         # Approximate bars per trading day
    
    # --- Auto-Scaled Parameters ---
    horizons: List[int]
    vol_window: int
    rolling_windows: List[int]
    tb_vertical_bars: int
    cv_folds: int
    max_depth: int
    n_estimators: int
    min_samples_leaf: int
    embargo_bars: int
    
    # --- Data Requirements ---
    min_bars_required: int
    ideal_bars: int
    # --- Walkforward Parameters ---
    wf_train_bars: int
    wf_test_bars: int
    wf_step_bars: int    
    # --- All Regime Options (for comparison runs) ---
    wf_configs: Dict[str, Dict[str, int]] = None
    
    def __post_init__(self):
        """Auto-generate walkforward configs for all regimes."""
        if self.wf_configs is None:
            self.wf_configs = {
                "reversion": {
                    "name": "Mean-Reversion",
                    "train_bars": int(self.bars_per_day * 21),
                    "test_bars": int(self.bars_per_day * 5),
                    "step_bars": int(self.bars_per_day * 5),
                },
                "balanced": {
                    "name": "Balanced",
                    "train_bars": int(self.bars_per_day * 63),
                    "test_bars": int(self.bars_per_day * 10),
                    "step_bars": int(self.bars_per_day * 10),
                },
                "momentum": {
                    "name": "Momentum",
                    "train_bars": int(self.bars_per_day * 126),
                    "test_bars": int(self.bars_per_day * 21),
                    "step_bars": int(self.bars_per_day * 21),
                },
            }
    
    def set_active_regime(self, regime: str):
        """Switch active walkforward config to a regime."""
        cfg = self.wf_configs[regime]
        self.wf_train_bars = cfg["train_bars"]
        self.wf_test_bars = cfg["test_bars"]
        self.wf_step_bars = cfg["step_bars"]
        print(f"✓ Active regime: {cfg['name']}")
# ============================================================
# TIMEFRAME CONFIGURATION FUNCTION
def get_timeframe_config(timeframe: str, total_bars: int = None) -> TimeframeConfig:
    """
    Generate configuration for any timeframe.
    
    Timeframe formats:
        Ticks:   "1T", "10T", "100T", "1000T"
        Seconds: "1S", "5S", "10S", "15S", "30S", "45S"
        Minutes: "1M", "2M", "3M", "5M", "10M", "15M", "30M", "45M"
        Hours:   "1H", "2H", "3H", "4H"
        Days:    "1D", "1W", "1MO", "3MO", "6MO", "12MO"
        Range:   "1R", "10R", "100R", "1000R"
    """
    
    # --- Base Reference: 1 Hour ---
    # All scaling is relative to 1H (approx 7 bars/day for futures)
    
    CONFIGS = {
        # ============ TICKS ============
        "1T": {
            "bars_per_day": 50000,
            "scale": 0.0001,  # Relative to 1H
        },
        "10T": {
            "bars_per_day": 5000,
            "scale": 0.001,
        },
        "100T": {
            "bars_per_day": 500,
            "scale": 0.01,
        },
        "1000T": {
            "bars_per_day": 50,
            "scale": 0.1,
        },
        
        # ============ SECONDS ============
        "1S": {
            "bars_per_day": 25200,  # 7 hours * 3600
            "scale": 0.0003,
        },
        "5S": {
            "bars_per_day": 5040,
            "scale": 0.0014,
        },
        "10S": {
            "bars_per_day": 2520,
            "scale": 0.003,
        },
        "15S": {
            "bars_per_day": 1680,
            "scale": 0.004,
        },
        "30S": {
            "bars_per_day": 840,
            "scale": 0.008,
        },
        "45S": {
            "bars_per_day": 560,
            "scale": 0.013,
        },
        
        # ============ MINUTES ============
        "1M": {
            "bars_per_day": 420,  # 7 hours * 60
            "scale": 0.017,
        },
        "2M": {
            "bars_per_day": 210,
            "scale": 0.033,
        },
        "3M": {
            "bars_per_day": 140,
            "scale": 0.05,
        },
        "5M": {
            "bars_per_day": 84,
            "scale": 0.083,
        },
        "10M": {
            "bars_per_day": 42,
            "scale": 0.17,
        },
        "15M": {
            "bars_per_day": 28,
            "scale": 0.25,
        },
        "30M": {
            "bars_per_day": 14,
            "scale": 0.5,
        },
        "45M": {
            "bars_per_day": 9.3,
            "scale": 0.75,
        },
        
        # ============ HOURS ============
        "1H": {
            "bars_per_day": 7,  # Base reference
            "scale": 1.0,
        },
        "2H": {
            "bars_per_day": 3.5,
            "scale": 2.0,
        },
        "3H": {
            "bars_per_day": 2.3,
            "scale": 3.0,
        },
        "4H": {
            "bars_per_day": 1.75,
            "scale": 4.0,
        },
        
        # ============ DAYS ============
        "1D": {
            "bars_per_day": 1,
            "scale": 7.0,
        },
        "1W": {
            "bars_per_day": 0.2,  # 1 bar per week
            "scale": 35.0,
        },
        "1MO": {
            "bars_per_day": 0.05,  # ~1 bar per month
            "scale": 140.0,
        },
        "3MO": {
            "bars_per_day": 0.016,
            "scale": 420.0,
        },
        "6MO": {
            "bars_per_day": 0.008,
            "scale": 840.0,
        },
        "12MO": {
            "bars_per_day": 0.004,
            "scale": 1680.0,
        },
        
        # ============ RANGE BARS ============
        "1R": {
            "bars_per_day": 1000,  # Highly variable
            "scale": 0.007,
        },
        "10R": {
            "bars_per_day": 100,
            "scale": 0.07,
        },
        "100R": {
            "bars_per_day": 10,
            "scale": 0.7,
        },
        "1000R": {
            "bars_per_day": 1,
            "scale": 7.0,
        },
    }
    
    if timeframe not in CONFIGS:
        raise ValueError(f"Unknown timeframe: {timeframe}. Valid: {list(CONFIGS.keys())}")
    
    cfg = CONFIGS[timeframe]
    scale = cfg["scale"]
    bars_per_day = cfg["bars_per_day"]
    
    # --- Scale Parameters (base = 1H values) ---
    
    # Horizons: scale inversely (more bars = shorter horizons in bar terms)
    base_horizons = [5, 10, 20, 50, 100]
    horizons = [max(1, int(h / scale)) for h in base_horizons]
    # Cap and dedupe
    horizons = sorted(set([min(h, 500) for h in horizons]))[:5]
    
    # Volatility window: ~1 week of data
    vol_window = max(5, min(200, int(50 / scale)))
    
    # Rolling windows
    base_rolling = [10, 20, 50, 100]
    rolling_windows = sorted(set([max(3, int(r / scale)) for r in base_rolling]))[:4]
    
    # Triple barrier vertical (target ~1 day)
    tb_vertical_bars = max(1, int(bars_per_day))
    
    # CV folds: fewer with less data
    if bars_per_day < 1:  # Daily or longer
        cv_folds = 3
    elif bars_per_day < 10:
        cv_folds = 4
    else:
        cv_folds = 5
    
    # Model complexity: simpler for low-frequency
    if scale > 10:  # Weekly or longer
        max_depth = 3
        n_estimators = 50
        min_samples_leaf = 10
    elif scale > 1:  # Daily to 4H
        max_depth = 4
        n_estimators = 100
        min_samples_leaf = 5
    else:  # Intraday
        max_depth = 6
        n_estimators = 200
        min_samples_leaf = 3
    
    # Embargo: ~1 day worth of bars
    embargo_bars = max(1, int(bars_per_day))
    
    # Data requirements
    min_bars_required = max(200, int(500 / scale))
    ideal_bars = max(1000, int(2000 / scale))

    # Walkforward: ~3 months train, ~1 week test
    wf_train_bars = int(bars_per_day * 63)   # ~63 trading days
    wf_test_bars = int(bars_per_day * 5)     # ~5 trading days
    wf_step_bars = wf_test_bars              # Non-overlapping by default
    
    return TimeframeConfig(
        timeframe=timeframe,
        bars_per_day=bars_per_day,
        horizons=horizons,
        vol_window=vol_window,
        rolling_windows=rolling_windows,
        tb_vertical_bars=tb_vertical_bars,
        cv_folds=cv_folds,
        max_depth=max_depth,
        n_estimators=n_estimators,
        min_samples_leaf=min_samples_leaf,
        embargo_bars=embargo_bars,
        min_bars_required=min_bars_required,
        ideal_bars=ideal_bars,
        wf_train_bars=wf_train_bars,
        wf_test_bars=wf_test_bars,
        wf_step_bars=wf_step_bars,
    )


def print_timeframe_config(cfg: TimeframeConfig, total_bars: int = None):
    """Pretty print the configuration."""
    print("=" * 60)
    print(f"TIMEFRAME CONFIGURATION: {cfg.timeframe}")
    print("=" * 60)
    print(f"\n{'Parameter':<25} {'Value':<35}")
    print("-" * 60)
    print(f"{'Bars per day':<25} {cfg.bars_per_day:<35}")
    print(f"{'Horizons':<25} {str(cfg.horizons):<35}")
    print(f"{'Volatility window':<25} {cfg.vol_window:<35}")
    print(f"{'Rolling windows':<25} {str(cfg.rolling_windows):<35}")
    print(f"{'TB vertical bars':<25} {cfg.tb_vertical_bars:<35}")
    print(f"{'CV folds':<25} {cfg.cv_folds:<35}")
    print(f"{'Max tree depth':<25} {cfg.max_depth:<35}")
    print(f"{'N estimators':<25} {cfg.n_estimators:<35}")
    print(f"{'Min samples leaf':<25} {cfg.min_samples_leaf:<35}")
    print(f"{'Embargo bars':<25} {cfg.embargo_bars:<35}")
    print(f"{'Min bars required':<25} {cfg.min_bars_required:<35}")
    print(f"{'Ideal bars':<25} {cfg.ideal_bars:<35}")
    print(f"{'WF train bars':<25} {cfg.wf_train_bars:<35}")
    print(f"{'WF test bars':<25} {cfg.wf_test_bars:<35}")
    print(f"{'WF step bars':<25} {cfg.wf_step_bars:<35}")
    if total_bars:
            status = "✓ OK" if total_bars >= cfg.min_bars_required else "⚠️ LOW"
            print(f"\n{'Your data':<25} {total_bars} bars {status}")
        
    print("=" * 60)
    
def print_walkforward_configs(cfg: TimeframeConfig):
    """Pretty print all walkforward regime options."""
    print("\n" + "=" * 60)
    print("WALKFORWARD REGIME OPTIONS")
    print("=" * 60)
    
    for regime, wf in cfg.wf_configs.items():
        active = " ← ACTIVE" if wf["train_bars"] == cfg.wf_train_bars else ""
        print(f"\n📊 {wf['name']}{active}")
        print(f"   Train: {wf['train_bars']:,} bars ({wf['train_bars']/cfg.bars_per_day:.0f} days)")
        print(f"   Test:  {wf['test_bars']:,} bars ({wf['test_bars']/cfg.bars_per_day:.0f} days)")
        print(f"   Step:  {wf['step_bars']:,} bars")
    
    print("=" * 60)


# ============================================================
# SET YOUR TIMEFRAME HERE * Manual Config, must match data file
# ============================================================

TIMEFRAME = "15M"  # ← Change this to match your data

# Auto-generate configuration (confirmation only)
TF_CONFIG = get_timeframe_config(TIMEFRAME)

# Preview (optional - comment out after confirming)
print_timeframe_config(TF_CONFIG)
print_walkforward_configs(TF_CONFIG)  

# ============================================================
# EXPORT PARAMETERS FOR USE THROUGHOUT NOTEBOOK
# ============================================================

HORIZONS = TF_CONFIG.horizons
VOL_WINDOW = TF_CONFIG.vol_window
ROLLING_WINDOWS = TF_CONFIG.rolling_windows
TB_VERTICAL_BARS = TF_CONFIG.tb_vertical_bars
CV_FOLDS = TF_CONFIG.cv_folds
MAX_DEPTH = TF_CONFIG.max_depth
N_ESTIMATORS = TF_CONFIG.n_estimators
MIN_SAMPLES_LEAF = TF_CONFIG.min_samples_leaf
EMBARGO_BARS = TF_CONFIG.embargo_bars
MIN_BARS_REQUIRED = TF_CONFIG.min_bars_required
EMBARGO_BARS = TF_CONFIG.embargo_bars
MIN_BARS_REQUIRED = TF_CONFIG.min_bars_required

# --- Walkforward Parameters ---
WF_TRAIN_BARS = TF_CONFIG.wf_train_bars
WF_TEST_BARS = TF_CONFIG.wf_test_bars
WF_STEP_BARS = TF_CONFIG.wf_step_bars
WF_CONFIGS = TF_CONFIG.wf_configs  # All regime options

logger.info(f"Timeframe config loaded: {TIMEFRAME}")

logger.info(
    f"Horizons: {HORIZONS}, Vol window: {VOL_WINDOW}, CV folds: {CV_FOLDS}, Max depth: {MAX_DEPTH}, N estimators: {N_ESTIMATORS}, Min samples leaf: {MIN_SAMPLES_LEAF}, Embargo bars: {EMBARGO_BARS}, Min bars required: {MIN_BARS_REQUIRED}, WF train bars: {WF_TRAIN_BARS}, WF test bars: {WF_TEST_BARS}, WF step bars: {WF_STEP_BARS}, Walkforward regimes: {list(WF_CONFIGS.keys())}, Total bars required: {MIN_BARS_REQUIRED}, Total bars ideal: {TF_CONFIG.ideal_bars}.")
````

````python
# ============================================================
# SECTION 01.9 — Data Timeframe Inspection & UNIX Translation
# ============================================================

import pandas as pd
import numpy as np
from datetime import datetime, timedelta

def inspect_timeframe(df, time_col=None):
    """
    Inspect CSV data for timeframe, date range, and UNIX timestamp handling.
    
    Automatically detects:
    - UNIX timestamps (seconds, milliseconds, microseconds, nanoseconds)
    - ISO datetime strings
    - Various date formats
    """
    
    print("=" * 70)
    print("DATA TIMEFRAME INSPECTION")
    print("=" * 70)
    
    # --- 1. Identify time column ---
    if time_col is None:
        # Try common names
        candidates = ['time', 'Time', 'timestamp', 'Timestamp', 'date', 'Date', 
                      'datetime', 'Datetime', 'DateTime', 'unix', 'Unix', 'UNIX']
        
        # Check index first
        if isinstance(df.index, pd.DatetimeIndex):
            time_col = "INDEX"
            time_data = df.index
        else:
            # Check first column (often UNIX timestamp)
            first_col = df.columns[0]
            if df[first_col].dtype in ['int64', 'float64']:
                time_col = first_col
                time_data = df[first_col]
            else:
                # Search for named column
                for c in candidates:
                    if c in df.columns:
                        time_col = c
                        time_data = df[time_col]
                        break
                else:
                    print("⚠️  Could not identify time column!")
                    print(f"   Columns: {df.columns.tolist()[:10]}...")
                    return None
    else:
        time_data = df.index if time_col == "INDEX" else df[time_col]
    
    print(f"\n📍 Time Column: '{time_col}'")
    print(f"   Data Type: {time_data.dtype}")
    print(f"   Sample Values (first 3): {time_data[:3].tolist()}")
    
    # --- 2. Detect and convert UNIX timestamps ---
    is_unix = False
    unix_unit = None
    
    if time_data.dtype in ['int64', 'float64'] and time_col != "INDEX":
        sample = time_data[0]
        
        # Detect UNIX unit based on magnitude
        if sample > 1e18:  # Nanoseconds
            unix_unit = 'ns'
            is_unix = True
        elif sample > 1e15:  # Microseconds
            unix_unit = 'us'
            is_unix = True
        elif sample > 1e12:  # Milliseconds
            unix_unit = 'ms'
            is_unix = True
        elif sample > 1e9:  # Seconds (after year 2001)
            unix_unit = 's'
            is_unix = True
        
        if is_unix:
            print(f"\n🔄 UNIX Timestamp Detected: {unix_unit}")
            time_data = pd.to_datetime(time_data, unit=unix_unit, utc=True)
            print(f"   Converted Sample: {time_data[0]}")
    
    # Convert to datetime if string
    if time_data.dtype == 'object':
        try:
            time_data = pd.to_datetime(time_data, utc=True)
            print(f"\n🔄 Parsed datetime strings")
        except:
            print(f"\n⚠️  Could not parse time data as datetime")
            return None
    
    # --- 3. Calculate date range ---
    start_time = time_data.min()
    end_time = time_data.max()
    total_span = end_time - start_time
    
    print(f"\n📅 DATE RANGE:")
    print(f"   Start:    {start_time}")
    print(f"   End:      {end_time}")
    print(f"   Span:     {total_span}")
    print(f"   Days:     {total_span.days}")
    
    # --- 4. Calculate bar frequency ---
    time_diffs = time_data.diff().dropna()

    # Handle both Series and TimedeltaIndex
    if hasattr(time_diffs, 'dt'):
        diff_seconds = time_diffs.dt.total_seconds()
    else:
        diff_seconds = time_diffs.total_seconds()

    # Convert to Series for consistent .mode() behavior
    diff_seconds = pd.Series(diff_seconds)

    mode_seconds = diff_seconds.mode()[0] if len(diff_seconds.mode()) > 0 else diff_seconds.median()
    median_seconds = diff_seconds.median()
    mean_seconds = diff_seconds.mean()

    
    def seconds_to_timeframe(seconds):
        """Convert seconds to human-readable timeframe."""
        if seconds < 1:
            return f"{int(seconds * 1000)}ms"
        elif seconds < 60:
            return f"{int(seconds)}S"
        elif seconds < 3600:
            mins = seconds / 60
            if mins == int(mins):
                return f"{int(mins)}M"
            return f"{mins:.1f}M"
        elif seconds < 86400:
            hours = seconds / 3600
            if hours == int(hours):
                return f"{int(hours)}H"
            return f"{hours:.1f}H"
        elif seconds < 604800:
            days = seconds / 86400
            if days == int(days):
                return f"{int(days)}D"
            return f"{days:.1f}D"
        elif seconds < 2592000:
            weeks = seconds / 604800
            return f"{int(weeks)}W"
        else:
            months = seconds / 2592000
            return f"{int(months)}MO"
    
    detected_tf = seconds_to_timeframe(mode_seconds)
    
    print(f"\n⏱️  BAR FREQUENCY:")
    print(f"   Mode (most common): {seconds_to_timeframe(mode_seconds)} ({mode_seconds:.0f} seconds)")
    print(f"   Median:             {seconds_to_timeframe(median_seconds)} ({median_seconds:.0f} seconds)")
    print(f"   Mean:               {seconds_to_timeframe(mean_seconds)} ({mean_seconds:.0f} seconds)")
    
    # Check for irregular intervals
    std_seconds = diff_seconds.std()
    cv = std_seconds / mean_seconds if mean_seconds > 0 else 0
    
    if cv > 0.1:
        print(f"\n   ⚠️  IRREGULAR INTERVALS DETECTED (CV={cv:.2f})")
        print(f"   Min gap: {seconds_to_timeframe(diff_seconds.min())}")
        print(f"   Max gap: {seconds_to_timeframe(diff_seconds.max())}")
    else:
        print(f"\n   ✓ Regular intervals (CV={cv:.4f})")
    
    # --- 5. Data summary ---
    print(f"\n📊 DATA SUMMARY:")
    print(f"   Total Bars:     {len(df):,}")
    print(f"   Bars per Day:   {len(df) / max(total_span.days, 1):.1f}")
    
    # Trading days estimation
    if median_seconds < 86400:  # Intraday
        if hasattr(time_data, 'dt'):
            unique_days = time_data.dt.date.nunique()
    else:
        unique_days = len(set(time_data.date))
        print(f"   Trading Days:   {unique_days}")
        print(f"   Avg Bars/Day:   {len(df) / unique_days:.1f}")
    
    # --- 6. Suggested timeframe config ---
    print(f"\n🎯 SUGGESTED TIMEFRAME CONFIG:")
    print(f"   TIMEFRAME = \"{detected_tf}\"")
    
    # --- 7. UNIX Conversion Helper ---
    print(f"\n" + "=" * 70)
    print("UNIX TIMESTAMP CONVERTER")
    print("=" * 70)
    
    if is_unix:
        print(f"\nYour data uses UNIX timestamps in {unix_unit} (unit='{unix_unit}')")
        print(f"\nTo convert throughout notebook, add after loading CSV:")
        print(f"""
    # Convert UNIX timestamp to datetime
    df_raw['{time_col}'] = pd.to_datetime(df_raw['{time_col}'], unit='{unix_unit}', utc=True)
    df_raw = df_raw.set_index('{time_col}')
        """)
    else:
        print("\n✓ Data already in datetime format")
        print(f"\n   First timestamp: {time_data[0]}")
        print(f"   Last timestamp:  {time_data[-1]}")
        
        # If you want to show what the original UNIX would have been:
        print(f"\n   (As UNIX seconds):")
        print(f"   First: {int(time_data[0].timestamp())}")
        print(f"   Last:  {int(time_data[-1].timestamp())}")

    
    print("=" * 70)
    
    # Return useful info
    return {
        "time_col": time_col,
        "start": start_time,
        "end": end_time,
        "span_days": total_span.days,
        "total_bars": len(df),
        "detected_timeframe": detected_tf,
        "bar_seconds": mode_seconds,
        "is_unix": is_unix,
        "unix_unit": unix_unit,
        "is_regular": cv < 0.1,
    }
# =============================================================================
# EXPECTED GAP FILTER (Weekends & Session Breaks)
# =============================================================================

def calculate_adjusted_cv(df, time_col=None):
    """
    Calculate CV excluding expected market gaps (weekends, holidays).
    For futures: Globex closes Fri 5PM ET → Sun 6PM ET
    """
    if time_col:
        time_data = pd.to_datetime(df[time_col])
    else:
        time_data = df.index
    
    diff = time_data.diff()
    if hasattr(diff, 'dt'):
        diff_seconds = diff.dt.total_seconds().dropna()
    else:
        diff_seconds = pd.Series(diff.total_seconds()).dropna()
    
    # Detect the "normal" bar interval (mode)
    normal_gap = diff_seconds.mode()[0]
    
    # Define maximum expected gap threshold
    # For 1H bars: normal_gap = 3600s, allow up to ~3x for session breaks
    # Weekend gaps (~50+ hours) will be excluded
    max_expected_gap = normal_gap * 4  # Adjust multiplier as needed
    
    # Filter to only "normal" trading gaps
    trading_gaps = diff_seconds[diff_seconds <= max_expected_gap]
    excluded_gaps = diff_seconds[diff_seconds > max_expected_gap]
    
    # Recalculate CV on trading gaps only
    mean_gap = trading_gaps.mean()
    std_gap = trading_gaps.std()
    adjusted_cv = std_gap / mean_gap if mean_gap > 0 else 0
    
    print("=" * 70)
    print("ADJUSTED CV ANALYSIS (Excluding Expected Gaps)")
    print("=" * 70)
    print(f"\n   Normal bar interval: {normal_gap/3600:.1f} hours")
    print(f"   Max expected gap:    {max_expected_gap/3600:.1f} hours")
    print(f"\n   Total intervals:     {len(diff_seconds)}")
    print(f"   Trading intervals:   {len(trading_gaps)}")
    print(f"   Excluded gaps:       {len(excluded_gaps)}")
    
    if len(excluded_gaps) > 0:
        print(f"\n   Excluded gap sizes:")
        for gap in sorted(excluded_gaps.unique()):
            count = (excluded_gaps == gap).sum()
            print(f"      {gap/3600:.1f}h ({gap/86400:.1f}d) × {count}")
    
    print(f"\n   Raw CV:              {diff_seconds.std()/diff_seconds.mean():.2f}")
    print(f"   Adjusted CV:         {adjusted_cv:.4f}")
    
    if adjusted_cv > 0.1:
        print(f"\n   ⚠️  IRREGULAR TRADING INTERVALS (CV={adjusted_cv:.2f})")
        print(f"      Check for missing bars within sessions")
    else:
        print(f"\n   ✓ Regular trading intervals")
    
    print("=" * 70)
    
    return {
        "raw_cv": diff_seconds.std()/diff_seconds.mean(),
        "adjusted_cv": adjusted_cv,
        "normal_gap_seconds": normal_gap,
        "excluded_gap_count": len(excluded_gaps),
        "is_regular": adjusted_cv < 0.1
    }
# ============================================================
# RUN INSPECTION
# ============================================================

# Inspect your loaded data (df_raw or df_prices)
timeframe_info = inspect_timeframe(df_raw)

# Optional: Auto-set TIMEFRAME based on detection
if timeframe_info:
    DETECTED_TIMEFRAME = timeframe_info["detected_timeframe"]
    print(f"\n💡 Auto-detected: TIMEFRAME = \"{DETECTED_TIMEFRAME}\"")
    print(f"   Update Section 00 config if this matches your intent.")
# Calculate adjusted CV excluding expected gaps
cv_info = calculate_adjusted_cv(df_raw)

# ------------------------------------------------------------
# VALIDATION CHECK — Section 01
# ------------------------------------------------------------
quick_check(df_prices, "df_prices")
print(f"Date range: {df_prices.index[0]} to {df_prices.index[-1]}")
````

````python
# ============================================================
# SECTION 03 — Label Engineering (Full Suite)
# ============================================================

logger.info("Starting Section 03 — Label Engineering (Full Suite)")

close = df_prices["Close"]
high = df_prices["High"]
low = df_prices["Low"]
open_ = df_prices["Open"]

# ------------------------------------------------------------
# 1. Triple-Barrier Labeling
# ------------------------------------------------------------
logger.info("Computing triple-barrier labels...")

tb_events = lbl_tb.generate_triple_barrier_events(
    df_prices,
    pt_sl=[1.5, 1.5],
    min_ret=0.001,
    horizon=TB_VERTICAL_BARS * 3600,  # Uses TF_CONFIG
    vol_span=VOL_WINDOW,
)
df_tb = lbl_tb.apply_triple_barrier(df_prices, tb_events)
df_tb = df_tb.rename(columns={"label": "tb_label"})

logger.info(f"Triple-barrier labels generated: {df_tb.shape}")

# ------------------------------------------------------------
# 2. Multi-Horizon Direction Labels
# ------------------------------------------------------------
logger.info("Computing multi-horizon labels...")

effective_timeframe = globals().get("DETECTED_TIMEFRAME") or TIMEFRAME
if effective_timeframe != TIMEFRAME:
    logger.warning(f"Using detected timeframe {effective_timeframe} instead of TIMEFRAME={TIMEFRAME} for horizon labels.")
if "get_timeframe_config" in globals():
    _tf = get_timeframe_config(effective_timeframe, total_bars=len(df_prices))
    horizons = _tf.horizons
else:
    horizons = HORIZONS
mh_labels = lbl_mh.build_horizon_direction_matrix(close, horizons=horizons)

logger.info(f"Multi-horizon labels generated for {len(horizons)} horizons")

# ------------------------------------------------------------
# 3. Forward Returns (Raw & Log)
# ------------------------------------------------------------
logger.info("Computing forward returns...")

fwd_returns = {}
for h in horizons:
    fwd_returns[f"fwd_ret_{h}"] = lbl_fwd.forward_return(close, horizon=h)
    fwd_returns[f"fwd_ret_log_{h}"] = lbl_fwd.forward_return(close, horizon=h, log=True)

df_fwd_returns = pd.DataFrame(fwd_returns, index=close.index)

# Simple forward direction
fwd_dir = lbl_fwd.directional_forward_return(close, horizon=horizons[2]).rename("fwd_dir")

logger.info(f"Forward returns computed: {df_fwd_returns.shape}")

# ------------------------------------------------------------
# 4. Volatility-Adjusted Labels
# ------------------------------------------------------------
logger.info("Computing volatility-adjusted labels...")

vol_adj_labels = {}

# Vol-normalized return
vol_adj_labels["vol_norm_ret"] = lbl_voladj.vol_normalized_return(close, horizon=horizons[2], vol_window=VOL_WINDOW)

# Z-scored forward return
vol_adj_labels["zscore_ret"] = lbl_voladj.zscored_forward_return(close, horizon=horizons[2], lookback=VOL_WINDOW)

# Risk-adjusted direction (Sharpe-like)
vol_adj_labels["risk_adj_dir"] = lbl_voladj.risk_adjusted_direction(close, horizon=horizons[2], vol_window=VOL_WINDOW)

# Scaled vol return (SVR target)
vol_adj_labels["svr_target"] = lbl_voladj.svr_target(close, horizon=horizons[2], vol_window=VOL_WINDOW)

# Drawdown-adjusted return
vol_adj_labels["dd_adj_ret"] = lbl_voladj.drawdown_adjusted_return(close, horizon=horizons[2])

df_vol_adj = pd.DataFrame(vol_adj_labels, index=close.index)

logger.info(f"Volatility-adjusted labels computed: {df_vol_adj.shape}")

# ------------------------------------------------------------
# 5. Volatility Features (for labels context)
# ------------------------------------------------------------
logger.info("Computing volatility measures...")

vol_measures = {}

vol_measures["vol_ewma"] = lbl_vol.ewma_volatility(close, span=VOL_WINDOW * 2)
vol_measures["vol_parkinson"] = lbl_vol.parkinson_vol(high, low, window=ROLLING_WINDOWS[1])
vol_measures["vol_garman_klass"] = lbl_vol.garman_klass_vol(open_, high, low, close, window=ROLLING_WINDOWS[1])
vol_measures["vol_yang_zhang"] = lbl_vol.yang_zhang_vol(open_, high, low, close, window=ROLLING_WINDOWS[1])
vol_measures["vol_regime"] = lbl_vol.volatility_regime(vol_measures["vol_ewma"], bins=4)

df_vol_measures = pd.DataFrame(vol_measures, index=close.index)

logger.info(f"Volatility measures computed: {df_vol_measures.shape}")

# ------------------------------------------------------------
# 6. Advanced Labels (from labels.py)
# ------------------------------------------------------------
logger.info("Computing advanced labels...")

adv_labels = {}

# Trend label
adv_labels["trend_label"] = lbl_main.apply_trend_label(df_prices, horizon_bars=horizons[2])

# Vol-adjusted trend
adv_labels["vol_trend_label"] = lbl_main.apply_vol_adjusted_trend_label(df_prices, horizon_bars=horizons[2], vol_window=VOL_WINDOW)

# Regime-conditioned label (if regime column exists later, update)
try:
    adv_labels["regime_label"] = lbl_main.apply_regime_conditioned_label(df_prices, horizon_bars=horizons[2])
except Exception as e:
    logger.warning(f"Regime label skipped: {e}")
    adv_labels["regime_label"] = np.nan

df_adv_labels = pd.DataFrame(adv_labels, index=close.index)

logger.info(f"Advanced labels computed: {df_adv_labels.shape}")

# ------------------------------------------------------------
# 7. Ensemble Label (Consensus)
# ------------------------------------------------------------
logger.info("Computing ensemble label...")

# Horizon consensus
horizon_cols = [f"dir_h{h}" for h in horizons]
mh_labels["horizon_consensus"] = np.sign(mh_labels[horizon_cols].sum(axis=1))
mh_labels["horizon_agreement"] = mh_labels[horizon_cols].apply(
    lambda row: (row == row.mode()[0]).sum() / len(row) if len(row.mode()) > 0 else 0, axis=1
)

# Multi-source ensemble
ensemble_sources = pd.DataFrame({
    "tb": df_tb["tb_label"].reindex(close.index).fillna(0),
    "fwd": fwd_dir.reindex(close.index).fillna(0),
    "trend": df_adv_labels["trend_label"].fillna(0),
    "horizon_cons": mh_labels["horizon_consensus"].reindex(close.index).fillna(0),
})

ensemble_sources["ensemble_vote"] = np.sign(ensemble_sources.sum(axis=1))
ensemble_sources["ensemble_strength"] = ensemble_sources[["tb", "fwd", "trend", "horizon_cons"]].sum(axis=1)
ensemble_sources["ensemble_agreement"] = (
    ensemble_sources[["tb", "fwd", "trend", "horizon_cons"]]
    .apply(lambda row: (row == row.mode()[0]).sum() if len(row.mode()) > 0 else 0, axis=1) / 4
)

logger.info("Ensemble label computed")

# ------------------------------------------------------------
# 8. Assemble Full Label Matrix
# ------------------------------------------------------------
logger.info("Assembling full label matrix...")

df_labels = pd.concat([
    df_tb,                    # t1, tb_label
    mh_labels,                # dir_h5..h100, horizon_consensus, horizon_agreement
    df_fwd_returns,           # fwd_ret_5..50, fwd_ret_log_5..50
    fwd_dir,                  # fwd_dir
    df_vol_adj,               # vol_norm_ret, zscore_ret, risk_adj_dir, svr_target, dd_adj_ret
    df_vol_measures,          # vol_ewma, vol_parkinson, vol_garman_klass, vol_yang_zhang, vol_regime
    df_adv_labels,            # trend_label, vol_trend_label, regime_label
    ensemble_sources[["ensemble_vote", "ensemble_strength", "ensemble_agreement"]],
], axis=1)

# Align to price index
df_labels = df_labels.reindex(df_prices.index)
df_labels["tb_label"] = df_labels["tb_label"].fillna(0)

logger.info(f"Final label matrix: {df_labels.shape}")
logger.info(f"Label columns ({len(df_labels.columns)}): {df_labels.columns.tolist()}")

# ------------------------------------------------------------
# 9. Label Summary Statistics
# ------------------------------------------------------------
print("\n" + "=" * 60)
print("LABEL MATRIX SUMMARY")
print("=" * 60)
print(f"\nShape: {df_labels.shape}")
print(f"\nNaN counts:")
print(df_labels.isna().sum().sort_values(ascending=False).head(10))
print(f"\nLabel distributions:")
for col in ["tb_label", "fwd_dir", "trend_label", "ensemble_vote", "horizon_consensus"]:
    if col in df_labels.columns:
        print(f"\n{col}:")
        print(df_labels[col].value_counts().sort_index())
print("=" * 60)
# ------------------------------------------------------------
# VALIDATION CHECK — Section 03
quick_check(df_labels, "Labels (df_labels)")
print(f"Label columns: {df_labels.columns.tolist()}")
print(f"Label value counts:\n{df_labels['tb_label'].value_counts()}")

# ------------------------------------------------------------
# 10. Save Label Matrix
# ------------------------------------------------------------
LABEL_FILE = LABEL_SAVE_DIR / "labels.parquet"
df_labels.to_parquet(LABEL_FILE)

logger.info(f"Saved label matrix to {LABEL_FILE}")
logger.info("Section 03 complete — Full label suite engineered.")
````

````python
# ============================================================
# DIAGNOSTIC #2, 03-04: ENSEMBLE LABEL FRAMEWORK
# ============================================================

def create_ensemble_label(df_labels, weights=None):
    """
    Create ensemble label from multiple label sources.
    
    Sources: tb_label, fwd_dir, trend_label, horizon_consensus
    """
    sources = ["tb_label", "fwd_dir", "trend_label", "horizon_consensus"]
    available = [s for s in sources if s in df_labels.columns]
    
    if weights is None:
        weights = {s: 1.0 for s in available}
    
    # Weighted vote
    weighted_sum = sum(
        df_labels[s].fillna(0) * weights.get(s, 1.0) 
        for s in available
    )
    
    # Normalize and sign
    ensemble = np.sign(weighted_sum)
    
    # Confidence = agreement level
    agreement = df_labels[available].apply(
        lambda row: (row == np.sign(row.sum())).sum() / len(row), axis=1
    )
    
    return ensemble, agreement

# Create with custom weights (favor triple-barrier and trend)
df_labels["ensemble_label"], df_labels["ensemble_conf"] = create_ensemble_label(
    df_labels,
    weights={"tb_label": 2.0, "fwd_dir": 1.0, "trend_label": 1.5, "horizon_consensus": 1.0}
)

# Optional: Use ensemble_label as alternative target
# y_ensemble = df_all["ensemble_label"]
````

````python
# ============================================================
# SECTION 04 — Dataset Assembly & Alignment + Multiclass Evaluation
# ============================================================

logger.info("Starting Section 04 — Dataset Assembly & Alignment")

import numpy as np
import pandas as pd

from pathlib import Path

from sklearn.metrics import (
    accuracy_score,
    balanced_accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix,
    roc_auc_score,
    roc_curve,
    auc,
)
from sklearn.preprocessing import label_binarize

import matplotlib.pyplot as plt


# ------------------------------------------------------------
# 1. Align Features and Labels on Timestamp
# ------------------------------------------------------------

logger.info("Aligning feature matrix and label matrix...")

df_all = df_features.join(df_labels, how="inner")

logger.info(
    f"Post-join dataset shape: {df_all.shape} "
    f"(features: {df_features.shape}, labels: {df_labels.shape})"
)


# ------------------------------------------------------------
# 2. Remove Rows with Missing Values
# ------------------------------------------------------------

before = df_all.shape[0]
# Only drop NaN on feature columns and the target label, not auxiliary columns like t1
essential_cols = list(df_features.columns) + ["tb_label", "trend_label", "ensemble_vote", "horizon_consensus"]
df_all.dropna(subset=essential_cols, inplace=True)
after = df_all.shape[0]

logger.info(f"Dropped {before - after} rows with NaNs. Clean dataset size: {after}")


# ------------------------------------------------------------
# 3. Validate Sorted DatetimeIndex
# ------------------------------------------------------------

cv_purged.validate_datetime_index(df_all.index)
logger.info("Datetime index validated after assembly.")


# ------------------------------------------------------------
# 4. Split Into Feature Matrix (X) and Label Vector(s) (y)
# ------------------------------------------------------------

logger.info("Splitting dataset into X (features) and y (target)...")
# ============================================================
# MULTI-LABEL TESTING CONFIG
# ============================================================
LABELS_TO_TEST = ["tb_label", "trend_label", "fwd_dir", "ensemble_vote", "horizon_consensus", "direction_label" ]
RUN_MULTI_LABEL = True  # Set True to test all labels

# Default single label (used when RUN_MULTI_LABEL=False)
TARGET_LABEL = "tb_label"


X = df_all[df_features.columns]
y = df_all[TARGET_LABEL]
# ============================================================
# LABEL-DERIVED FEATURES (Feature Engineering from Labels)
# ============================================================

logger.info("Engineering label-derived features...")

# --- A. Lagged Labels (avoid leakage with shift) ---
LABEL_LAG = 1  # Minimum 1 to avoid leakage

label_features = {}

# Previous label signals
label_features["prev_tb_label"] = df_labels["tb_label"].shift(LABEL_LAG)
label_features["prev_fwd_dir"] = df_labels["fwd_dir"].shift(LABEL_LAG)
label_features["prev_trend_label"] = df_labels["trend_label"].shift(LABEL_LAG)

# --- B. Label Momentum (rolling signal consistency) ---
for window in [3, 5, 10]:
    label_features[f"label_momentum_{window}"] = (
        df_labels["tb_label"].shift(LABEL_LAG).rolling(window).mean()
    )
    label_features[f"label_consistency_{window}"] = (
        df_labels["tb_label"].shift(LABEL_LAG).rolling(window).apply(
            lambda x: (x == x.mode()[0]).sum() / len(x) if len(x.mode()) > 0 else 0
        )
    )

# --- C. Horizon Agreement Features ---
horizon_cols = sorted([c for c in df_labels.columns if c.startswith("dir_h")],
                      key=lambda x: int(x.split("dir_h")[1]) if x.split("dir_h")[1].isdigit() else 0)
horizon_cols_available = [c for c in horizon_cols if c in df_labels.columns]
if not horizon_cols_available:
    logger.warning("No horizon direction columns found in df_labels; setting horizon_strength=0.")
    label_features["horizon_strength"] = 0
else:
    label_features["horizon_strength"] = (
        df_labels[horizon_cols_available].shift(LABEL_LAG).sum(axis=1)
    )
if "horizon_agreement" in df_labels.columns:
    label_features["horizon_agreement"] = df_labels["horizon_agreement"].shift(LABEL_LAG)
else:
    logger.warning("horizon_agreement not found in df_labels; setting to 0.")
    label_features["horizon_agreement"] = 0
if "dir_h5" in df_labels.columns and "dir_h100" in df_labels.columns:
    label_features["short_long_divergence"] = (
        df_labels["dir_h5"].shift(LABEL_LAG) - df_labels["dir_h100"].shift(LABEL_LAG)
    )
else:
    logger.warning("dir_h5/dir_h100 missing; setting short_long_divergence=0.")
    label_features["short_long_divergence"] = 0
# --- D. Ensemble Signal Features ---
label_features["ensemble_strength"] = df_labels["ensemble_strength"].shift(LABEL_LAG)
label_features["ensemble_agreement"] = df_labels["ensemble_agreement"].shift(LABEL_LAG)

# --- E. Vol-Adjusted Signal Features ---
label_features["zscore_ret_lag"] = df_labels["zscore_ret"].shift(LABEL_LAG)
label_features["risk_adj_dir_lag"] = df_labels["risk_adj_dir"].shift(LABEL_LAG)
label_features["vol_regime"] = df_labels["vol_regime"].shift(LABEL_LAG)

# --- F. Label Change Detection ---
label_features["label_flip"] = (
    df_labels["tb_label"].shift(LABEL_LAG) != df_labels["tb_label"].shift(LABEL_LAG + 1)
).astype(int)
label_features["label_streak"] = (
    df_labels["tb_label"].shift(LABEL_LAG)
    .groupby((df_labels["tb_label"].shift(LABEL_LAG) != df_labels["tb_label"].shift(LABEL_LAG + 1)).cumsum())
    .cumcount() + 1
)

# --- G. Cross-Label Interactions ---
label_features["tb_x_trend"] = (
    df_labels["tb_label"].shift(LABEL_LAG) * df_labels["trend_label"].shift(LABEL_LAG)
)
label_features["tb_x_vol_regime"] = (
    df_labels["tb_label"].shift(LABEL_LAG) * df_labels["vol_regime"].shift(LABEL_LAG)
)

# Assemble and join
df_label_features = pd.DataFrame(label_features, index=df_labels.index)
df_label_features = df_label_features.add_prefix("LBL_")  # Prefix for identification

# Join to feature matrix
X = X.join(df_label_features, how="left")

# Checkup if new features joined
print(f"Label-derived features added: {len([c for c in X.columns if c.startswith('LBL_')])}")
print(f"Sample LBL columns: {[c for c in X.columns if c.startswith('LBL_')][:5]}")

logger.info(f"Added {len(df_label_features.columns)} label-derived features")
logger.info(f"New X shape: {X.shape}")

# Update feature list
label_feature_cols = df_label_features.columns.tolist()


# --- Label audit (print + log) ---
vc = y.value_counts(dropna=False)
unique_classes = sorted(pd.unique(y))
trade_rate_true = float((y != 0).mean())

print("\n=== ensemble_label distribution ===")
print(vc)
print("\nUnique classes:", unique_classes)
print("Trade rate (nonzero):", trade_rate_true)

logger.info("=== ensemble_label distribution ===\n" + vc.to_string())
logger.info(f"Unique classes: {unique_classes}")
logger.info(f"Trade rate (nonzero): {trade_rate_true:.4f}")

logger.info(f"Feature matrix X shape: {X.shape}")
logger.info(f"Target vector y shape: {y.shape}")


# ------------------------------------------------------------
# 5. Optional: Correlation Filtering (Feature Reduction)
# ------------------------------------------------------------

CORR_FILTER = False
CORR_THRESHOLD = 0.95

if CORR_FILTER:
    logger.info("Applying correlation filtering...")
    corr = X.corr().abs()
    upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
    to_drop = [col for col in upper.columns if any(upper[col] > CORR_THRESHOLD)]
    logger.info(f"Dropping {len(to_drop)} highly correlated features.")
    X = X.drop(columns=to_drop)

logger.info(f"Final X shape after correlation filtering: {X.shape}")


# ------------------------------------------------------------
# 6. Initial Chronological Train/Test Split (diagnostics only)
# ------------------------------------------------------------

logger.info("Performing initial chronological split...")

TRAIN_RATIO = 0.70
split_idx = int(len(X) * TRAIN_RATIO)

X_train_init = X.iloc[:split_idx]
X_test_init  = X.iloc[split_idx:]
y_train_init = y.iloc[:split_idx]
y_test_init  = y.iloc[split_idx:]

logger.info(f"Initial train set shape: {X_train_init.shape}")
logger.info(f"Initial test  set shape: {X_test_init.shape}")

# Ensure multiclass training/eval is meaningful
train_classes = pd.unique(y_train_init)
test_classes = pd.unique(y_test_init)
if len(train_classes) < 2:
    raise ValueError(f"Train split has <2 classes: {train_classes}")
if len(test_classes) < 2:
    logger.info(f"Warning: test split has <2 classes: {test_classes} (metrics may be limited)")


# ------------------------------------------------------------
# 7. Save Fully Assembled Dataset
# ------------------------------------------------------------

try:
    ROOT  # noqa: F821
except NameError:
    ROOT = Path.cwd()
    logger.info(f"ROOT not found; using cwd as ROOT: {ROOT}")

artifacts_dir = Path(ROOT) / "artifacts"
artifacts_dir.mkdir(parents=True, exist_ok=True)

DATASET_FILE = artifacts_dir / "full_dataset.parquet"
try:
    df_all.to_parquet(DATASET_FILE)
    logger.info(f"Saved fully assembled modeling dataset to {DATASET_FILE}")
except Exception as e:
    logger.info(f"Parquet save skipped: {e}")
# Optional fallback
    csv_file = artifacts_dir / "full_dataset.csv"
    df_all.to_csv(csv_file)
    logger.info(f"Saved dataset as CSV instead: {csv_file}")
# ------------------------------------------------------------
# VALIDATION CHECK — Section 04
quick_check(df_all, "Assembled Dataset (df_all)")
quick_check(X, "Features (X)")
quick_check(y, "Target (y)")
print(f"Feature groups: {list(DataValidator.detect_feature_groups(X).keys())}")
# ------------------------------------------------------------
# 8. Multiclass Evaluation (macro + weighted + report + ROC)
# ------------------------------------------------------------

RUN_SECTION04_EVAL = True  # set False to skip multiclass eval diagnostics
````

````python
# ============================================================
# SECTION DIAGNOSTIC #2, 05-06 : CV Splitter Validation
# ============================================================

logger.info("=" * 60)
logger.info("CV SPLITTER DIAGNOSTIC CHECK")
logger.info("=" * 60)

def diagnose_cv_splitter(cv, X, name):
    """Run diagnostics on a CV splitter."""
    results = {
        "name": name,
        "n_splits": 0,
        "avg_train_size": 0,
        "avg_test_size": 0,
        "min_train_size": float('inf'),
        "max_train_size": 0,
        "min_test_size": float('inf'),
        "max_test_size": 0,
        "overlap_detected": False,  # Start clean, only True if overlap found
        "errors": []
    }
    
    train_sizes = []
    test_sizes = []
    
    try:
        for fold_idx, (train_idx, test_idx) in enumerate(cv.split(X)):
            train_sizes.append(len(train_idx))
            test_sizes.append(len(test_idx))
            
            # Check for overlap (leakage)
            train_set = set(train_idx)
            test_set = set(test_idx)
            overlap = train_set & test_set
            if overlap:
                results["overlap_detected"] = True
                results["errors"].append(f"Fold {fold_idx}: {len(overlap)} overlapping indices!")
            
            results["n_splits"] += 1
            
            # Safety limit
            if results["n_splits"] >= 100:
                results["errors"].append("Stopped at 100 splits (limit reached)")
                break
                
    except Exception as e:
        results["errors"].append(f"Exception: {str(e)[:100]}")
        return results
    
    if train_sizes:
        results["avg_train_size"] = int(np.mean(train_sizes))
        results["avg_test_size"] = int(np.mean(test_sizes))
        results["min_train_size"] = min(train_sizes)
        results["max_train_size"] = max(train_sizes)
        results["min_test_size"] = min(test_sizes)
        results["max_test_size"] = max(test_sizes)
    
    return results

def print_cv_report(results):
    """Print a formatted report for a CV splitter."""
    status = "✓" if not results["errors"] and not results["overlap_detected"] else "✗"
    
    print(f"\n{status} {results['name']}")
    print("-" * 50)
    
    if results["errors"]:
        for err in results["errors"]:
            print(f"  ERROR: {err}")
    else:
        print(f"  Splits generated:    {results['n_splits']}")
        print(f"  Avg train size:      {results['avg_train_size']}")
        print(f"  Avg test size:       {results['avg_test_size']}")
        print(f"  Train size range:    [{results['min_train_size']}, {results['max_train_size']}]")
        print(f"  Test size range:     [{results['min_test_size']}, {results['max_test_size']}]")
        print(f"  Overlap (leakage):   {'YES ⚠️' if results['overlap_detected'] else 'None ✓'}")

# Run diagnostics on all CV splitters
cv_splitters = [
    (cv_purged_kfold, "PurgedKFold"),
    (cpcv_splitter, "CombinatorialPurgedKFold (CPCV)"),
    (cv_clustered, "ClusteredCV"),
    (cv_walkforward, "WalkForwardCV"),
]

print(f"\nDataset size: {len(df_all)} rows")
print(f"Event end times: {len(event_end_times)} values")
print(f"Event end times range: {event_end_times.min()} to {event_end_times.max()}")

all_passed = True
for cv, name in cv_splitters:
    results = diagnose_cv_splitter(cv, df_all, name)
    print_cv_report(results)
    if results["errors"] or results["overlap_detected"]:
        all_passed = False

print("\n" + "=" * 60)
if all_passed:
    print("✓ ALL CV SPLITTERS PASSED VALIDATION")
else:
    print("✗ SOME CV SPLITTERS HAVE ISSUES - CHECK ABOVE")
print("=" * 60)

logger.info("CV Splitter diagnostic complete.")
````

````python
# ============================================================
# SECTION 07 — Model Diagnostics & Cross-Validation Analysis
# ============================================================

logger.info("Starting Section 07 — Diagnostics & CV Analysis")

# ------------------------------------------------------------
# Define helper functions ONCE (outside the loop)
# ------------------------------------------------------------
import numpy as np
import pandas as pd
from sklearn.metrics import brier_score_loss
def get_rolling_window(n_rows: int, target_pct: float = 0.15, min_window: int = 15) -> int:
    """Dynamically compute rolling window size based on data length."""
    window = int(n_rows * target_pct)
    return max(min_window, window)


def brier_decomposition(y_true, y_prob, classes=None, class_labels=None, n_bins=10):
    """
    Unified Brier score decomposition for any number of classes.
    """
    y_true = np.asarray(y_true)
    
    if classes is None:
        classes = sorted(np.unique(y_true))
    
    n_classes = len(classes)
    n_samples = len(y_true)
    
    if isinstance(y_prob, pd.DataFrame):
        y_prob_matrix = y_prob.values
    elif isinstance(y_prob, np.ndarray):
        if y_prob.ndim == 1:
            if n_classes == 2:
                y_prob_matrix = np.column_stack([1 - y_prob, y_prob])
            else:
                raise ValueError("1D y_prob only valid for binary classification")
        else:
            y_prob_matrix = y_prob
    else:
        y_prob_matrix = np.asarray(y_prob)
    
    if y_prob_matrix.shape[1] != n_classes:
        raise ValueError(f"y_prob has {y_prob_matrix.shape[1]} columns but {n_classes} classes detected")
    
    y_onehot = np.zeros((n_samples, n_classes))
    for i, cls in enumerate(classes):
        y_onehot[:, i] = (y_true == cls).astype(int)
    
    class_results = {}
    
    for i, cls in enumerate(classes):
        y_binary = y_onehot[:, i]
        y_prob_cls = y_prob_matrix[:, i]
        y_mean = np.mean(y_binary)
        
        bins = np.linspace(0, 1, n_bins + 1)
        bin_indices = np.digitize(y_prob_cls, bins, right=True)
        bin_indices = np.clip(bin_indices, 1, n_bins) - 1
        
        reliability = 0.0
        resolution = 0.0
        
        for b in range(n_bins):
            mask = bin_indices == b
            n_k = np.sum(mask)
            if n_k > 0:
                p_k = np.mean(y_prob_cls[mask])
                o_k = np.mean(y_binary[mask])
                reliability += n_k * (p_k - o_k) ** 2
                resolution += n_k * (o_k - y_mean) ** 2
        
        reliability /= n_samples
        resolution /= n_samples
        uncertainty = y_mean * (1 - y_mean)
        brier = brier_score_loss(y_binary, y_prob_cls)
        
        label = class_labels.get(cls, str(cls)) if class_labels else str(cls)
        class_results[cls] = {
            'label': label, 'brier': brier, 'reliability': reliability,
            'resolution': resolution, 'uncertainty': uncertainty, 'base_rate': y_mean
        }
    
    multiclass_brier = np.mean(np.sum((y_prob_matrix - y_onehot) ** 2, axis=1))
    avg_reliability = np.mean([r['reliability'] for r in class_results.values()])
    avg_resolution = np.mean([r['resolution'] for r in class_results.values()])
    avg_uncertainty = np.mean([r['uncertainty'] for r in class_results.values()])
    
    p_clipped = np.clip(y_prob_matrix, 1e-10, 1 - 1e-10)
    sample_entropy = -np.sum(p_clipped * np.log2(p_clipped), axis=1)
    pred_entropy = np.mean(sample_entropy)
    max_entropy = np.log2(n_classes)
    normalized_entropy = pred_entropy / max_entropy if max_entropy > 0 else 0
    
    confidence = np.mean(np.max(y_prob_matrix, axis=1))
    pred_variance = np.mean(np.var(y_prob_matrix, axis=0))
    uniform_prob = 1.0 / n_classes
    sharpness = np.mean(np.abs(y_prob_matrix - uniform_prob))
    
    y_pred = np.array([classes[i] for i in np.argmax(y_prob_matrix, axis=1)])
    correct_mask = y_pred == y_true
    accuracy = np.mean(correct_mask)
    
    overconfidence = np.mean(np.max(y_prob_matrix[~correct_mask], axis=1)) if np.any(~correct_mask) else 0.0
    correct_confidence = np.mean(np.max(y_prob_matrix[correct_mask], axis=1)) if np.any(correct_mask) else 0.0
    
    print("=" * 60)
    print(f"BRIER SCORE DECOMPOSITION ({n_classes}-CLASS)")
    print("=" * 60)
    print("\n--- Per-Class Breakdown ---")
    print(f"{'Class':<12} {'Brier':>8} {'Reliab':>8} {'Resol':>8} {'Uncert':>8} {'Base%':>8}")
    print("-" * 60)
    for cls in classes:
        r = class_results[cls]
        print(f"{r['label']:<12} {r['brier']:>8.4f} {r['reliability']:>8.4f} "
              f"{r['resolution']:>8.4f} {r['uncertainty']:>8.4f} {r['base_rate']*100:>7.1f}%")
    print("-" * 60)
    print(f"{'Average':<12} {np.mean([r['brier'] for r in class_results.values()]):>8.4f} "
          f"{avg_reliability:>8.4f} {avg_resolution:>8.4f} {avg_uncertainty:>8.4f}")
    print("\n--- Multi-Class Total ---")
    print(f"Multi-class Brier Score: {multiclass_brier:.4f}  (lower is better)")
    print("\n--- Uncertainty Metrics ---")
    print(f"Prediction Entropy:      {pred_entropy:.4f} / {max_entropy:.4f}  ({normalized_entropy*100:.1f}% of max)")
    print(f"Mean Confidence:         {confidence:.4f}  (avg max probability)")
    print(f"Prediction Variance:     {pred_variance:.4f}  (spread of estimates)")
    print(f"Sharpness:               {sharpness:.4f}  (distance from uniform)")
    print("\n--- Calibration Diagnostics ---")
    print(f"Accuracy:                {accuracy:.4f}")
    print(f"Confidence when RIGHT:   {correct_confidence:.4f}")
    print(f"Confidence when WRONG:   {overconfidence:.4f}")
    if overconfidence > 0.7 and accuracy < 0.5:
        print(f"  ⚠ High overconfidence on errors - consider calibration")
    print("=" * 60)
    
    return {
        'n_classes': n_classes, 'classes': classes, 'per_class': class_results,
        'multiclass_brier': multiclass_brier, 'avg_reliability': avg_reliability,
        'avg_resolution': avg_resolution, 'avg_uncertainty': avg_uncertainty,
        'pred_entropy': pred_entropy, 'normalized_entropy': normalized_entropy,
        'confidence': confidence, 'pred_variance': pred_variance, 'sharpness': sharpness,
        'accuracy': accuracy, 'correct_confidence': correct_confidence, 'overconfidence': overconfidence
    }


# Storage for cross-label comparison
section07_results = {}

for current_label in labels_to_run:
    
    print(f"\n{'#'*70}")
    print(f"#  SECTION 07 DIAGNOSTICS: {current_label.upper()}")
    print(f"{'#'*70}\n")
    
    # Get this label's predictions from Section 06
    if RUN_MULTI_LABEL:
        if current_label not in all_results:
            logger.warning(f"Skipping {current_label} — no results from Section 06")
            continue
        df_predictions = all_results[current_label]["predictions"]
    
    # ------------------------------------------------------------
    # 0. Load and Clean Predictions
    # ------------------------------------------------------------
    
    df_preds = df_predictions.copy()
    df_preds["timestamp"] = df_preds.index
    
    prob_cols = [col for col in df_preds.columns if col.startswith("prob_class_")]
    df_preds_clean = df_preds.dropna(subset=["true"] + prob_cols)
    logger.info(f"[{current_label}] Cleaned predictions: {len(df_preds_clean)} rows (dropped {len(df_preds) - len(df_preds_clean)} NaN rows)")
    
    actual_classes = sorted(df_preds_clean["true"].unique())
    n_classes = len(actual_classes)
    
    window_size = get_rolling_window(len(df_preds_clean))
    logger.info(f"[{current_label}] Using rolling window size: {window_size}")
    
    df_preds_with_regime = df_preds_clean.join(X["Regime_Combo"])
    
    # ------------------------------------------------------------
    # 1. Confusion Matrix (Across All Folds)
    # ------------------------------------------------------------
    from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

    logger.info(f"[{current_label}] Generating overall confusion matrix...")
    
    cm = confusion_matrix(df_preds_clean["true"], df_preds_clean["pred"])
    disp = ConfusionMatrixDisplay(cm, display_labels=actual_classes)
    disp.plot(cmap="Blues")
    plt.title(f"Confusion Matrix — {current_label}")
    plt.savefig(ROOT / "artifacts" / f"confusion_matrix_{current_label}.png", dpi=150)
    plt.show()
    
    # ------------------------------------------------------------
    # 2. ROC AUC Score
    # ------------------------------------------------------------
    
    y_true = df_preds_clean["true"]
    
    if n_classes == 1:
        logger.warning(f"[{current_label}] Only 1 class present — cannot compute ROC AUC")
        roc_auc = np.nan
    elif n_classes == 2:
        pos_class = actual_classes[1]
        y_score = df_preds_clean[f"prob_class_{pos_class}"]
        roc_auc = roc_auc_score(y_true == pos_class, y_score)
        logger.info(f"[{current_label}] Binary ROC AUC: {roc_auc:.4f}")
    else:
        y_probs = df_preds_clean[prob_cols].values
        roc_auc = roc_auc_score(y_true, y_probs, multi_class='ovr', average='macro')
        logger.info(f"[{current_label}] Multi-class ROC AUC (macro): {roc_auc:.4f}")
    
    # ------------------------------------------------------------
    # 3. Precision-Recall Curves (Per Class)
    # ------------------------------------------------------------
    
    logger.info(f"[{current_label}] Plotting precision-recall curves...")
    from sklearn.metrics import precision_recall_curve

    fig, ax = plt.subplots(figsize=(10, 6))
    for cls in actual_classes:
        y_binary = (df_preds_clean["true"] == cls).astype(int)
        prec, rec, _ = precision_recall_curve(y_binary, df_preds_clean[f"prob_class_{cls}"])
        ax.plot(rec, prec, label=f"Class {cls}")
    ax.set_xlabel("Recall")
    ax.set_ylabel("Precision")
    ax.set_title(f"Precision-Recall Curves — {current_label}")
    ax.legend()
    plt.savefig(ROOT / "artifacts" / f"pr_curves_{current_label}.png", dpi=150)
    plt.show()
    
    # ------------------------------------------------------------
    # 4. Rolling Error Rate
    # ------------------------------------------------------------
    
    logger.info(f"[{current_label}] Computing rolling error rate...")
    
    df_preds_clean["error"] = df_preds_clean["true"] != df_preds_clean["pred"]
    
    plt.figure(figsize=(14, 4))
    plt.plot(df_preds_clean["error"].rolling(window_size).mean(), label=f"Rolling Error Rate ({window_size})")
    plt.xlabel("Date")
    plt.ylabel("Error Rate")
    plt.title(f"Rolling Error Rate Over Time — {current_label}")
    plt.legend()
    plt.savefig(ROOT / "artifacts" / f"rolling_error_{current_label}.png", dpi=150)
    plt.show()
    
    print(f"\n[{current_label}] Error column stats:")
    print(f"  Mean: {df_preds_clean['error'].mean():.4f}")
    print(f"  Sum: {df_preds_clean['error'].sum()}")
    print(f"  Total rows: {len(df_preds_clean)}")
    print(f"  Accuracy: {(df_preds_clean['true'] == df_preds_clean['pred']).mean():.4f}")
    
    # ------------------------------------------------------------
    # 5. Error Analysis by Regime
    # ------------------------------------------------------------
    
    logger.info(f"[{current_label}] Analyzing errors by regime...")
    
    df_preds_with_regime = df_preds_clean.join(X["Regime_Combo"], how="left")
    
    regime_errors = df_preds_with_regime.groupby("Regime_Combo").agg(
        error_rate=("error", "mean"),
        count=("error", "count")
    ).sort_values("error_rate", ascending=False)
    
    print(f"\n[{current_label}] Error Rate by Regime:")
    print(regime_errors.round(4))
    
    # ------------------------------------------------------------
    # 6. Per-Fold Performance Summary
    # ------------------------------------------------------------
    
    logger.info(f"[{current_label}] Computing per-fold performance...")
    
    fold_results = []
    
    if n_classes == 2:
        pos_class = actual_classes[1]
    
    if n_classes >= 2:
        for fold in df_preds_clean["fold"].unique():
            fold_df = df_preds_clean[df_preds_clean["fold"] == fold]
            acc = (fold_df["true"] == fold_df["pred"]).mean()
            
            if n_classes == 2:
                try:
                    auc = roc_auc_score(fold_df["true"] == pos_class, fold_df[f"prob_class_{pos_class}"])
                except:
                    auc = np.nan
            else:
                try:
                    auc = roc_auc_score(fold_df["true"], fold_df[prob_cols].values, multi_class='ovr', average='macro')
                except:
                    auc = np.nan
            
            fold_results.append({"fold": fold, "accuracy": acc, "roc_auc": auc, "n_samples": len(fold_df)})
        
        df_fold_summary = pd.DataFrame(fold_results)
        print(f"\n[{current_label}] Per-Fold Summary:")
        print(df_fold_summary.round(4))
        print(f"\nMean Accuracy: {df_fold_summary['accuracy'].mean():.4f} ± {df_fold_summary['accuracy'].std():.4f}")
        print(f"Mean ROC AUC:  {df_fold_summary['roc_auc'].mean():.4f} ± {df_fold_summary['roc_auc'].std():.4f}")
    
    # ------------------------------------------------------------
    # 7a. Feature Stability Across CV Folds
    # ------------------------------------------------------------
    
    logger.info(f"[{current_label}] Computing feature stability across folds...")
    
    feature_stability = {}
    
    for col in X.columns:
        corrs = []
        for fold in df_preds["fold"].unique():
            fold_df = df_preds[df_preds["fold"] == fold]
            fold_df = fold_df[~fold_df.index.duplicated(keep='first')]
            valid_idx = fold_df.index.intersection(X.index)
            if len(valid_idx) > 0:
                x_vals = X.loc[valid_idx, col].values
                prob_vals = fold_df.loc[valid_idx, prob_cols].max(axis=1).values
                if np.std(x_vals) > 0 and np.std(prob_vals) > 0:
                    corrs.append(np.corrcoef(x_vals, prob_vals)[0, 1])
        feature_stability[col] = np.nanmean(corrs) if corrs else np.nan
    # create df_stability from the dictionary
    df_stability = pd.DataFrame.from_dict(feature_stability, orient="index", columns=["stability"])
    df_stability.sort_values("stability", ascending=False, inplace=True)
    stable_features = pd.Series(feature_stability).sort_values(ascending=False)
    print(f"\n[{current_label}] Top 10 Stable Features Across Folds:")
    print(stable_features.head(10).round(4))
    
    # ------------------------------------------------------------
    # 7b. Probability Calibration Curves
    # ------------------------------------------------------------
    
    logger.info(f"[{current_label}] Computing calibration curves...")
    from sklearn.calibration import calibration_curve

    fig, axes = plt.subplots(1, n_classes, figsize=(5*n_classes, 5))
    if n_classes == 1:
        axes = [axes]
    
    for idx, cls in enumerate(actual_classes):
        ax = axes[idx]
        y_binary = (df_preds_clean["true"] == cls).astype(int)
        y_prob = df_preds_clean[f"prob_class_{cls}"]
        
        fraction_of_positives, mean_predicted_value = calibration_curve(
            y_binary, y_prob, n_bins=10, strategy='uniform'
        )
        
        ax.plot(mean_predicted_value, fraction_of_positives, marker='o', label=f"Class {cls}")
        ax.plot([0, 1], [0, 1], 'k--', label="Perfect Calibration")
        ax.set_xlabel("Mean Predicted Probability")
        ax.set_ylabel("Fraction of Positives")
        ax.set_title(f"Calibration — Class {cls}")
        ax.legend()
    
    plt.suptitle(f"Calibration Curves — {current_label}", fontsize=12, fontweight='bold')
    plt.tight_layout()
    plt.savefig(ROOT / "artifacts" / f"calibration_curves_{current_label}.png", dpi=150)
    plt.show()
    
    # ------------------------------------------------------------
    # 8. Per-Class Rolling AUC (Stability Over Time)
    # ------------------------------------------------------------
    
    logger.info(f"[{current_label}] Computing rolling AUC per class...")
    
    rolling_aucs = {}
    
    if n_classes >= 2:
        for cls in actual_classes:
            try:
                aucs = []
                for fold in df_preds_clean["fold"].unique():
                    idx = df_preds_clean[df_preds_clean["fold"] == fold].index
                    valid_idx = [i for i in idx if i in df_preds_clean.index]
                    if len(valid_idx) > 0 and f"prob_class_{cls}" in df_preds_clean.columns:
                        y_bin = (df_preds_clean.loc[valid_idx, "true"] == cls).astype(int)
                        y_prob = df_preds_clean.loc[valid_idx, f"prob_class_{cls}"]
                        if y_bin.nunique() == 2:
                            aucs.append(roc_auc_score(y_bin, y_prob))
                rolling_aucs[cls] = np.mean(aucs) if aucs else np.nan
            except Exception as e:
                logger.warning(f"[{current_label}] Could not compute AUC for class {cls}: {e}")
                rolling_aucs[cls] = np.nan
        
        print(f"\n[{current_label}] Mean Rolling AUC per Class:")
        for cls, auc in rolling_aucs.items():
            print(f"  Class {cls}: {auc:.4f}" if not np.isnan(auc) else f"  Class {cls}: N/A")
    
    # ------------------------------------------------------------
    # 9. Rolling Class Probabilities
    # ------------------------------------------------------------
    
    logger.info(f"[{current_label}] Plotting rolling class probabilities...")
    
    plt.figure(figsize=(14, 5))
    for cls in actual_classes:
        col = f"prob_class_{cls}"
        if col in df_preds_clean.columns:
            plt.plot(df_preds_clean[col].rolling(window_size).mean(), label=f"P(Class {cls})")
    plt.xlabel("Date")
    plt.ylabel("Mean Probability")
    plt.title(f"Rolling Mean Predicted Probabilities — {current_label}")
    plt.legend()
    plt.savefig(ROOT / "artifacts" / f"rolling_probs_{current_label}.png", dpi=150)
    plt.show()
    
    # ------------------------------------------------------------
    # 10. ROC Curves (Multi-Class)
    # ------------------------------------------------------------
    
    if n_classes >= 2:
        logger.info(f"[{current_label}] Plotting ROC curves...")
        
        fig, ax = plt.subplots(figsize=(8, 8))
        for cls in actual_classes:
            y_binary = (df_preds_clean["true"] == cls).astype(int)
            y_prob = df_preds_clean[f"prob_class_{cls}"]
            fpr, tpr, _ = roc_curve(y_binary, y_prob)
            auc_cls = roc_auc_score(y_binary, y_prob)
            ax.plot(fpr, tpr, label=f"Class {cls} (AUC={auc_cls:.3f})")
        ax.plot([0, 1], [0, 1], "k--", label="Random")
        ax.set_xlabel("False Positive Rate")
        ax.set_ylabel("True Positive Rate")
        ax.set_title(f"ROC Curves — {current_label}")
        ax.legend()
        plt.savefig(ROOT / "artifacts" / f"roc_curves_{current_label}.png", dpi=150)
        plt.show()
    
    # ------------------------------------------------------------
    # 11. Brier Score Decomposition
    # ------------------------------------------------------------
    
    logger.info(f"[{current_label}] Computing Brier score decomposition...")
    
    print(f"\n[{current_label}] BRIER SCORE ANALYSIS:")
    brier_prob_cols = [f"prob_class_{c}" for c in actual_classes]
    brier_results = brier_decomposition(
        df_preds_clean["true"],
        df_preds_clean[brier_prob_cols],
        classes=list(actual_classes),
        class_labels={-1.0: "Bearish", 0.0: "Neutral", 1.0: "Bullish"}
    )
    
    # ------------------------------------------------------------
    # Store summary for comparison
    # ------------------------------------------------------------
    
    section07_results[current_label] = {
        "accuracy": (df_preds_clean["true"] == df_preds_clean["pred"]).mean(),
        "roc_auc": roc_auc,
        "mean_error": df_preds_clean["error"].mean(),
        "n_samples": len(df_preds_clean),
        "n_classes": n_classes,
        "fold_summary": df_fold_summary if n_classes >= 2 else None,
        "regime_errors": regime_errors,
        "stable_features": stable_features,
        "rolling_aucs": rolling_aucs if n_classes >= 2 else {},
        "brier_results": brier_results,
    }
    
    logger.info(f"[{current_label}] Section 07 complete.")
    
    # Close all figures to free memory
    plt.close('all')

# ------------------------------------------------------------
# VALIDATION CHECK — Section 07
# ------------------------------------------------------------
quick_check(df_preds_clean, "Clean Predictions")
print(f"Bull col: {get_bull_col(df_preds_clean)}, Bear col: {get_bear_col(df_preds_clean)}")

if "Regime_Combo" in df_preds_with_regime.columns:
    validate_by_regime(df_preds_with_regime, regime_col="Regime_Combo", target_col="true", pred_col="pred")

logger.info("Section 07 Complete — All labels processed.")
````

````python
# ============================================================
# SECTION 08 — Walkforward Simulation
# ============================================================

logger.info("Starting Section 08 — Walkforward Simulation")

# Storage for cross-label comparison
section08_results = {}

for current_label in labels_to_run:
    
    print(f"\n{'#'*70}")
    print(f"#  SECTION 08 WALKFORWARD: {current_label.upper()}")
    print(f"{'#'*70}\n")
    
    # Set y for this label
    y = df_all[current_label]
    
    # Skip if too few classes
    if y.nunique() < 2:
        logger.warning(f"Skipping {current_label} — only {y.nunique()} class(es)")
        continue
    
    # ------------------------------------------------------------
    # 1. Walkforward CV Settings — Multi-Regime
    # ------------------------------------------------------------
    
    ALL_CLASSES = sorted(y.unique())
    logger.info(f"[{current_label}] Target classes: {ALL_CLASSES}")
    
    all_regime_results = {}
    
    for regime_name, regime_cfg in WF_CONFIGS.items():
        
        logger.info("=" * 60)
        logger.info(f"[{current_label}] WALKFORWARD REGIME: {regime_cfg['name'].upper()}")
        logger.info(f"  Train: {regime_cfg['train_bars']} bars, Test: {regime_cfg['test_bars']} bars")
        logger.info("=" * 60)
        
        cv_walkforward = cv_walk.RollingWindowCV(
            train_size=regime_cfg["train_bars"],
            test_size=regime_cfg["test_bars"],
            step=regime_cfg["step_bars"],
            event_end_times=event_end_times
        )
        
        # ------------------------------------------------------------
        # 2. Initialize Model for Walkforward
        # ------------------------------------------------------------
        
        wf_model = RandomForestClassifier(**MODEL_PARAMS)
        
        # ------------------------------------------------------------
        # 3. Walkforward Loop
        # ------------------------------------------------------------
        
        wf_predictions = []
        wf_metrics = []
        
        for w, (train_idx, test_idx) in enumerate(cv_walkforward.split(df_all)):
            
            logger.info(f"[{current_label}][{regime_name}] Window {w+1}: train={len(train_idx)}, test={len(test_idx)}")
            
            X_train = X.iloc[train_idx]
            y_train = y.iloc[train_idx]
            X_test = X.iloc[test_idx]
            y_test = y.iloc[test_idx]
            
            wf_model.fit(X_train, y_train)
            
            wf_probs = wf_model.predict_proba(X_test)
            wf_preds = wf_model.predict(X_test)
            model_classes = list(wf_model.classes_)
            
            acc = accuracy_score(y_test, wf_preds)
            f1_macro = f1_score(y_test, wf_preds, average='macro', zero_division=0)
            f1_weighted = f1_score(y_test, wf_preds, average='weighted', zero_division=0)
            
            try:
                test_classes = set(np.unique(y_test))
                trained_classes = set(model_classes)
                
                if len(test_classes) >= 2 and test_classes.issubset(trained_classes):
                    if len(test_classes) == len(trained_classes):
                        roc = roc_auc_score(y_test, wf_probs, multi_class='ovr', average='macro')
                    elif len(sorted(test_classes)) == 2:
                        pos_class = sorted(test_classes)[-1]
                        pos_idx = model_classes.index(pos_class)
                        roc = roc_auc_score((y_test == pos_class).astype(int), wf_probs[:, pos_idx])
                    else:
                        roc = np.nan
                else:
                    roc = np.nan
            except Exception:
                roc = np.nan
            
            wf_metrics.append({
                "regime": regime_name,
                "window": w,
                "accuracy": acc,
                "f1_macro": f1_macro,
                "f1_weighted": f1_weighted,
                "roc_auc": roc,
                "n_train": len(train_idx),
                "n_test": len(test_idx)
            })
            
            pred_dict = {
                "timestamp": X_test.index,
                "true": y_test.values,
                "pred": wf_preds,
                "window": w
            }
            
            for cls in ALL_CLASSES:
                if cls in model_classes:
                    cls_idx = model_classes.index(cls)
                    pred_dict[f"prob_class_{cls}"] = wf_probs[:, cls_idx]
                else:
                    pred_dict[f"prob_class_{cls}"] = np.zeros(len(y_test))
            
            wf_pred_df = pd.DataFrame(pred_dict).set_index("timestamp")
            wf_predictions.append(wf_pred_df)
        
        # Skip regime if dataset too short for a single train+test window
        if not wf_predictions:
            logger.warning(
                f"[{current_label}][{regime_name}] Skipped — "
                f"{len(df_all)} bars < {regime_cfg['train_bars'] + regime_cfg['test_bars']} needed"
            )
            continue

        # Store regime results
        df_wf_preds = pd.concat(wf_predictions).sort_index()
        df_wf_preds["regime"] = regime_name
        df_wf_metrics = pd.DataFrame(wf_metrics)
        
        all_regime_results[regime_name] = {
            "predictions": df_wf_preds,
            "metrics": df_wf_metrics,
        }
        
        logger.info(f"[{current_label}][{regime_name}] Complete — {len(df_wf_preds)} predictions")
    
    # ------------------------------------------------------------
    # 4. Consolidate All Regimes
    # ------------------------------------------------------------
    
    if not all_regime_results:
        logger.warning(f"[{current_label}] No regime fit the dataset — skipping label")
        continue
    
    df_all_preds = pd.concat([r["predictions"] for r in all_regime_results.values()])
    df_all_metrics = pd.concat([r["metrics"] for r in all_regime_results.values()])
    
    logger.info(f"[{current_label}] Total predictions: {len(df_all_preds)}")
    
    # ------------------------------------------------------------
    # 5. Save Artifacts
    # ------------------------------------------------------------
    
    for regime_name, results in all_regime_results.items():
        pred_file = ROOT / "artifacts" / f"walkforward_predictions_{current_label}_{regime_name}.parquet"
        metrics_file = ROOT / "artifacts" / f"walkforward_metrics_{current_label}_{regime_name}.parquet"
        results["predictions"].to_parquet(pred_file)
        results["metrics"].to_parquet(metrics_file)
    
    df_all_preds.to_parquet(ROOT / "artifacts" / f"walkforward_predictions_{current_label}_all.parquet")
    df_all_metrics.to_parquet(ROOT / "artifacts" / f"walkforward_metrics_{current_label}_all.parquet")
    
    logger.info(f"[{current_label}] Saved walkforward artifacts.")
    
    # ------------------------------------------------------------
    # 6. Walkforward Diagnostics & Plots
    # ------------------------------------------------------------
    
    # 6A — REGIME COMPARISON
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    fig.suptitle(f"Walkforward Regime Comparison — {current_label}", fontsize=14, fontweight='bold')
    
    for regime_name, results in all_regime_results.items():
        df_m = results["metrics"]
        roc_valid = df_m[["window", "roc_auc"]].dropna()
        axes[0].plot(roc_valid["window"], roc_valid["roc_auc"], marker='o', label=regime_name, alpha=0.8)
        axes[1].plot(df_m["window"], df_m["f1_macro"], marker='o', label=regime_name, alpha=0.8)
        axes[2].plot(df_m["window"], df_m["accuracy"], marker='o', label=regime_name, alpha=0.8)
    
    axes[0].set_title("ROC AUC")
    axes[1].set_title("F1 Macro")
    axes[2].set_title("Accuracy")
    
    for ax in axes:
        ax.set_xlabel("Window")
        ax.legend()
        ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(ROOT / "artifacts" / f"walkforward_regime_comparison_{current_label}.png", dpi=150)
    plt.show()
    
    # Summary table
    summary = df_all_metrics.groupby("regime")[["roc_auc", "f1_macro", "accuracy"]].agg(['mean', 'std'])
    print(f"\n[{current_label}] REGIME COMPARISON — Mean Metrics:")
    print(summary.round(4))
    
    # 6B — PER-REGIME DETAIL PLOTS
    for regime_name, results in all_regime_results.items():
        
        df_wf_preds = results["predictions"].copy()
        df_wf_metrics = results["metrics"]
        
        print(f"\n[{current_label}] DETAILED PLOTS: {regime_name.upper()}")
        
        # Window-level metrics
        plt.figure(figsize=(15, 5))
        plt.plot(df_wf_metrics["window"], df_wf_metrics["roc_auc"].interpolate(), label="ROC AUC", linewidth=2)
        plt.plot(df_wf_metrics["window"], df_wf_metrics["f1_macro"], label="F1 Macro", linewidth=2)
        plt.plot(df_wf_metrics["window"], df_wf_metrics["accuracy"], label="Accuracy", linewidth=2)
        plt.title(f"[{current_label}][{regime_name}] Performance by Window")
        plt.xlabel("Window")
        plt.ylabel("Score")
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.savefig(ROOT / "artifacts" / f"walkforward_window_perf_{current_label}_{regime_name}.png", dpi=150)
        plt.show()
        
        # Prediction drift
        wf_window = get_rolling_window(len(df_wf_preds))
        
        plt.figure(figsize=(15, 5))
        for cls in ALL_CLASSES:
            col = f"prob_class_{cls}"
            if col in df_wf_preds.columns:
                rolling_prob = df_wf_preds[col].rolling(wf_window, min_periods=1).mean()
                plt.plot(rolling_prob, label=f"P(Class {cls})", linewidth=1.5)
        plt.title(f"[{current_label}][{regime_name}] Prediction Drift")
        plt.ylabel("Mean Probability")
        plt.xlabel("Time")
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.savefig(ROOT / "artifacts" / f"walkforward_drift_{current_label}_{regime_name}.png", dpi=150)
        plt.show()
        
        # Rolling error
        df_wf_preds["error"] = df_wf_preds["true"] != df_wf_preds["pred"]
        
        fig, axes = plt.subplots(1, 2, figsize=(15, 5))
        fig.suptitle(f"[{current_label}][{regime_name}] Error Analysis", fontsize=12)
        
        axes[0].plot(df_wf_preds["error"].rolling(wf_window, min_periods=1).mean())
        axes[0].set_title("Rolling Error Rate (Overall)")
        axes[0].set_ylabel("Error Rate")
        axes[0].grid(True, alpha=0.3)
        
        for cls in ALL_CLASSES:
            mask = df_wf_preds["true"] == cls
            if mask.sum() > 0:
                class_errors = df_wf_preds.loc[mask, "error"].rolling(wf_window // 3, min_periods=1).mean()
                axes[1].plot(class_errors, label=f"Class {cls}", alpha=0.8)
        axes[1].set_title("Rolling Error Rate (Per Class)")
        axes[1].legend()
        axes[1].grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig(ROOT / "artifacts" / f"walkforward_error_{current_label}_{regime_name}.png", dpi=150)
        plt.show()
    
    # 6C — Regime breakdown
    df_all_preds_reg = df_all_preds.join(X["Regime_Combo"])
    
    wf_regime_scores = df_all_preds_reg.groupby("Regime_Combo").apply(
        lambda g: pd.Series({
            "accuracy": accuracy_score(g["true"], g["pred"]),
            "f1_macro": f1_score(g["true"], g["pred"], average="macro", zero_division=0),
            "count": len(g),
        })
    )
    
    fig, ax = plt.subplots(figsize=(12, 6))
    wf_regime_scores[["accuracy", "f1_macro"]].plot(kind="bar", ax=ax, width=0.8)
    plt.title(f"Walkforward by Market Regime — {current_label}")
    plt.ylabel("Score")
    plt.xlabel("Regime")
    plt.xticks(rotation=45, ha='right')
    plt.legend(loc='lower right')
    plt.tight_layout()
    plt.savefig(ROOT / "artifacts" / f"walkforward_regime_breakdown_{current_label}.png", dpi=150)
    plt.show()
    
    print(f"\n[{current_label}] WALKFORWARD METRICS BY REGIME:")
    print(wf_regime_scores.round(4))
    
    # ------------------------------------------------------------
    # Store summary for comparison
    # ------------------------------------------------------------
    
    section08_results[current_label] = {
        "mean_accuracy": df_all_metrics["accuracy"].mean(),
        "mean_f1_macro": df_all_metrics["f1_macro"].mean(),
        "mean_roc_auc": df_all_metrics["roc_auc"].mean(),
        "std_accuracy": df_all_metrics["accuracy"].std(),
        "n_windows": len(df_all_metrics),
        "n_predictions": len(df_all_preds),
        "regime_results": all_regime_results,
        "regime_scores": wf_regime_scores,
    }
    
    logger.info(f"[{current_label}] Section 08 complete.")
    
    # Close all figures
    plt.close('all')

logger.info("Section 08 Complete — All labels processed.")
````

````python
# ============================================================
# SECTION 09 — Backtesting Engine & Execution Simulation
# ============================================================

logger.info("Starting Section 09 — Backtesting Engine")


# ------------------------------------------------------------
# 1. Load Walkforward Predictions
# ------------------------------------------------------------

df_wf = df_wf_preds.copy()

logger.info(f"Loaded walkforward prediction set: {df_wf.shape[0]} rows.")


# ------------------------------------------------------------
# 2. Signal Generation (Multiple Strategies)
# ------------------------------------------------------------

logger.info("Generating signals for multiple strategies...")

# Strategy 1: Direct prediction (aggressive, always-in)
df_wf["signal_direct"] = df_wf["pred"]


# Strategy 2: Confidence threshold (conservative, selective)
CONF_THRESH = 0.55

def generate_signal(row, thresh=0.5):
    """Dynamic signal generation based on available classes."""
    prob_cols = [c for c in row.index if c.startswith("prob_class_")]
    if not prob_cols:
        return 0
    
    # Find max probability class
    max_col = max(prob_cols, key=lambda c: row[c])
    max_prob = row[max_col]
    max_class = int(float(max_col.replace("prob_class_", "")))
    
    if max_prob > thresh and max_class != 0:
        return max_class  # Returns 1 or -1
    return 0

df_wf["signal_thresh"] = df_wf.apply(generate_signal, axis=1)

# Strategy 3: Net probability (directional momentum)

# For net_prob, use bullish - bearish (handle 2-class case)
bull_col = get_bull_col(df_wf)
bear_col = get_bear_col(df_wf)

if bull_col and bear_col:
    df_wf["net_prob"] = df_wf[bull_col] - df_wf[bear_col]
elif bull_col:
    df_wf["net_prob"] = df_wf[bull_col]
else:
    df_wf["net_prob"] = 0
    print("WARNING: No probability columns found for net_prob strategy!")

NET_THRESH = 0.25

def signal_net(net_prob, thresh=NET_THRESH):
    if net_prob > thresh:
        return 1
    elif net_prob < -thresh:
        return -1
    return 0

df_wf["signal_net"] = df_wf["net_prob"].apply(signal_net)

# Summary
logger.info(f"Signal distribution (direct): {df_wf['signal_direct'].value_counts().to_dict()}")
logger.info(f"Signal distribution (threshold): {df_wf['signal_thresh'].value_counts().to_dict()}")
logger.info(f"Signal distribution (net prob): {df_wf['signal_net'].value_counts().to_dict()}")


# ------------------------------------------------------------
# 3. Load Market Returns for PnL Computation
# ------------------------------------------------------------

logger.info("Merging market returns into walkforward predictions...")

df_wf = df_wf.join(df_prices["returns"], how="left", rsuffix="_mkt")
df_wf["returns"] = df_wf["returns"].fillna(0)

df_wf["price"] = df_prices["Close"].reindex(df_wf.index).ffill()

logger.info("Market data merged.")


# ------------------------------------------------------------
# 4. Backtest Configuration
# ------------------------------------------------------------

POSITION_SIZE = 1.0   # per trade, scalable
POINT_VALUE = 20      # point per dollar value // SUBJECT TO INSTRUMENT
SLIPPAGE = 25.0        # $2 per executed contract
COMMISSION = 2.50      # $2.5 per contract

logger.info(f"Backtest config: POSITION_SIZE={POSITION_SIZE}, POINT_VALUE={POINT_VALUE}, SLIPPAGE={SLIPPAGE}, COMMISSION={COMMISSION}")


# ------------------------------------------------------------
# 5. Backtest All Strategies
# ------------------------------------------------------------

strategies = {
    "direct": "signal_direct",
    "threshold": "signal_thresh", 
    "net_prob": "signal_net"
}

performance_results = {}

for strat_name, signal_col in strategies.items():
    
    logger.info(f"Backtesting strategy: {strat_name}")
    
    # Execution delay (1-bar)
    df_wf[f"{strat_name}_exec"] = df_wf[signal_col].shift(1).fillna(0)
    
    # Position sizing
    df_wf[f"{strat_name}_position"] = df_wf[f"{strat_name}_exec"] * POSITION_SIZE
    
    # PnL computation
    df_wf[f"{strat_name}_pnl"] = (
        df_wf[f"{strat_name}_position"] * 
        df_wf["returns"] * 
        df_wf["price"] * 
        POINT_VALUE
    )
    
    # Transaction costs (applied when position changes)
    df_wf[f"{strat_name}_trade"] = df_wf[f"{strat_name}_exec"].diff().fillna(0).abs()
    df_wf[f"{strat_name}_cost"] = df_wf[f"{strat_name}_trade"] * (SLIPPAGE + COMMISSION)
    
    # Net PnL
    df_wf[f"{strat_name}_pnl_net"] = df_wf[f"{strat_name}_pnl"] - df_wf[f"{strat_name}_cost"]
    
    # Equity curve
    df_wf[f"{strat_name}_equity"] = df_wf[f"{strat_name}_pnl_net"].cumsum()
    
    # Max drawdown
    rolling_max = df_wf[f"{strat_name}_equity"].cummax()
    df_wf[f"{strat_name}_drawdown"] = df_wf[f"{strat_name}_equity"] - rolling_max
    max_dd = df_wf[f"{strat_name}_drawdown"].min()
    
    # Performance stats
    daily_pnl = df_wf[f"{strat_name}_pnl_net"].resample("1D").sum()
    
    sharpe = (np.sqrt(252) * daily_pnl.mean() / daily_pnl.std()) if daily_pnl.std() != 0 else 0
    sortino = (np.sqrt(252) * daily_pnl.mean() / daily_pnl[daily_pnl < 0].std()) if daily_pnl[daily_pnl < 0].std() != 0 else 0
    
    performance_results[strat_name] = {
        "Total PnL": df_wf[f"{strat_name}_pnl_net"].sum(),
        "Sharpe Ratio": sharpe,
        "Sortino Ratio": sortino,
        "Max Drawdown": max_dd,
        "Win Rate": (df_wf[f"{strat_name}_pnl_net"] > 0).mean(),
        "Num Trades": df_wf[f"{strat_name}_trade"].sum(),
    }

logger.info("All strategies backtested.")


# ------------------------------------------------------------
# 6. Performance Comparison
# ------------------------------------------------------------

df_performance = pd.DataFrame(performance_results).T
logger.info(f"Strategy comparison:\n{df_performance}")

# Display as formatted table
print("\n" + "="*60)
print("STRATEGY PERFORMANCE COMPARISON")
print("="*60)
print(df_performance.to_string())
print("="*60 + "\n")


# ------------------------------------------------------------
# 7. Equity Curves (All Strategies)
# ------------------------------------------------------------

plt.figure(figsize=(14, 6))
for strat_name in strategies.keys():
    plt.plot(df_wf[f"{strat_name}_equity"], label=f"{strat_name.title()}")
plt.title("Equity Curves — All Strategies")
plt.xlabel("Time")
plt.ylabel("Equity ($)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()


# ------------------------------------------------------------
# 8. Drawdown Curves (All Strategies)
# ------------------------------------------------------------
# Check drawdown values
print("=== Drawdown values ===")
for strat_name in strategies.keys():
    dd = df_wf[f"{strat_name}_drawdown"]
    print(f"{strat_name}: min={dd.min():.2f}, max={dd.max():.2f}, NaN={dd.isna().sum()}")

# Plot each drawdown separately
fig, axes = plt.subplots(3, 1, figsize=(14, 10))
for i, strat_name in enumerate(strategies.keys()):
    axes[i].plot(df_wf[f"{strat_name}_drawdown"])
    axes[i].set_title(f"Drawdown — {strat_name.title()}")
    axes[i].set_ylabel("Drawdown ($)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 5))
for strat_name in strategies.keys():
    plt.plot(df_wf[f"{strat_name}_drawdown"], label=f"{strat_name.title()}")
plt.title("Drawdown Curves — All Strategies")
plt.xlabel("Time")
plt.ylabel("Drawdown ($)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(df_wf["direct_pnl_net"].describe())
print(df_wf["threshold_pnl_net"].describe())
print(df_wf["net_prob_pnl_net"].describe())

# ------------------------------------------------------------
# 9. Signal Distribution Comparison
# ------------------------------------------------------------

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for i, (strat_name, signal_col) in enumerate(strategies.items()):
    df_wf[signal_col].value_counts().sort_index().plot(kind="bar", ax=axes[i])
    axes[i].set_title(f"Signal Distribution — {strat_name.title()}")
    axes[i].set_xlabel("Signal")
    axes[i].set_ylabel("Count")
plt.tight_layout()
plt.show()

# Debug: Check if data exists for each strategy
print("=== DEBUG: Checking strategy data ===")
for strat_name in strategies.keys():
    equity_col = f"{strat_name}_equity"
    dd_col = f"{strat_name}_drawdown"
    pnl_col = f"{strat_name}_pnl_net"
    
    print(f"\n{strat_name}:")
    print(f"  Equity - min: {df_wf[equity_col].min():.2f}, max: {df_wf[equity_col].max():.2f}, NaN: {df_wf[equity_col].isna().sum()}")
    print(f"  Drawdown - min: {df_wf[dd_col].min():.2f}, max: {df_wf[dd_col].max():.2f}, NaN: {df_wf[dd_col].isna().sum()}")
    print(f"  PnL Net - sum: {df_wf[pnl_col].sum():.2f}, NaN: {df_wf[pnl_col].isna().sum()}")

# Debug: Check daily resampling for rolling Sharpe
print("\n=== DEBUG: Daily PnL resampling ===")
for strat_name in strategies.keys():
    daily_pnl = df_wf[f"{strat_name}_pnl_net"].resample("1D").sum()
    print(f"{strat_name}: {len(daily_pnl)} days, NaN: {daily_pnl.isna().sum()}, zeros: {(daily_pnl == 0).sum()}")


# ------------------------------------------------------------
# 10. Rolling Sharpe (20-day window)
# ------------------------------------------------------------
# Check date range for rolling Sharpe
print(f"\n=== Date range ===")
print(f"Start: {df_wf.index.min()}")
print(f"End: {df_wf.index.max()}")
print(f"Total days: {(df_wf.index.max() - df_wf.index.min()).days}")

ROLLING_WINDOW = horizons[2] if "horizons" in globals() else HORIZONS[2] 

plt.figure(figsize=(14, 5))
for strat_name in strategies.keys():
    daily_pnl = df_wf[f"{strat_name}_pnl_net"].resample("1D").sum()
    rolling_sharpe = (
        daily_pnl.rolling(ROLLING_WINDOW, min_periods=5).mean() / 
        daily_pnl.rolling(ROLLING_WINDOW, min_periods=5).std()
    ) * np.sqrt(252)
    plt.plot(rolling_sharpe, label=f"{strat_name.title()}")
plt.title("Rolling Sharpe Ratio (60-day window)")
plt.xlabel("Time")
plt.ylabel("Sharpe Ratio")
plt.axhline(y=0, color='red', linestyle='--', alpha=0.5)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"Data spans: {df_wf.index.min()} to {df_wf.index.max()}")
print(f"Total days: {(df_wf.index.max() - df_wf.index.min()).days}")
# ============================================================
# MULTI-LABEL ENSEMBLE FILTERING
# ============================================================

if USE_MULTILABEL_FILTER:
    logger.info(f"Applying multi-label filter (threshold={LABEL_AGREEMENT_THRESHOLD})...")
    
    # Get label agreement from df_labels
    label_cols = ["tb_label", "fwd_dir", "trend_label", "horizon_consensus"]
    
    # Reindex to match backtest DataFrame
    label_agreement = pd.DataFrame(index=df_wf.index)
    
    for col in label_cols:
        if col in df_labels.columns:
            label_agreement[col] = df_labels[col].reindex(df_wf.index).fillna(0)
    
    # Count how many labels agree with the primary signal (tb_label)
    primary = label_agreement["tb_label"]
    label_agreement["agree_count"] = (
        (label_agreement["fwd_dir"] == primary).astype(int) +
        (label_agreement["trend_label"] == primary).astype(int) +
        (label_agreement["horizon_consensus"] == primary).astype(int) +
        1  # tb_label agrees with itself
    )
    
    # Filter: only trade when agreement >= threshold
    filter_mask = label_agreement["agree_count"] >= LABEL_AGREEMENT_THRESHOLD
    
    # Apply filter to signals
    df_wf["signal_unfiltered"] = df_wf["signal_direct"].copy()
    df_wf["signal_filtered"] = np.where(filter_mask, df_wf["signal_unfiltered"], 0)
    df_wf["label_agreement"] = label_agreement["agree_count"]
    
    # Stats
    n_original = (df_wf["signal_unfiltered"] != 0).sum()
    n_filtered = (df_wf["signal_filtered"] != 0).sum()
    
    logger.info(f"Multi-label filter: {n_original} → {n_filtered} trades ({n_filtered/n_original*100:.1f}% kept)")
    
    # Use filtered signal for subsequent backtest
    ACTIVE_SIGNAL_COL = "signal_filtered"
else:
    ACTIVE_SIGNAL_COL = "signal_direct"
    logger.info("Multi-label filter: DISABLED")
# ============================================================
# REGIME-AWARE POSITION SIZING
# ============================================================

if USE_REGIME_SIZING:
    logger.info("Applying regime-aware position sizing...")
    
    # Define regime multipliers (customize based on regime_performance analysis)
    # Higher multiplier = larger position in that regime
    REGIME_SIZE_MULTIPLIERS = {
        # Profitable regimes → full or increased size
        # Losing regimes → reduced or zero size
    }
    
    # Auto-generate from regime performance if available
    if 'REGIME_PERFORMANCE' in dir():
        for regime in REGIME_PERFORMANCE.index:
            sharpe = REGIME_PERFORMANCE.loc[regime, "Sharpe"]
            if sharpe > 1.0:
                REGIME_SIZE_MULTIPLIERS[regime] = 1.5  # Increase size
            elif sharpe > 0.5:
                REGIME_SIZE_MULTIPLIERS[regime] = 1.0  # Normal size
            elif sharpe > 0:
                REGIME_SIZE_MULTIPLIERS[regime] = 0.5  # Reduced size
            else:
                REGIME_SIZE_MULTIPLIERS[regime] = 0.0  # Skip this regime
    
    # Apply multipliers
    df_wf["regime"] = X["Regime_Combo"].reindex(df_wf.index)
    df_wf["size_multiplier"] = df_wf["regime"].map(REGIME_SIZE_MULTIPLIERS).fillna(1.0)
    
    # Adjust position
    df_wf["position_base"] = df_wf[ACTIVE_SIGNAL_COL] * POSITION_SIZE
    df_wf["position_regime_adj"] = df_wf["position_base"] * df_wf["size_multiplier"]
    
    # Stats
    reduced_pct = (df_wf["size_multiplier"] < 1.0).mean() * 100
    skipped_pct = (df_wf["size_multiplier"] == 0).mean() * 100
    
    logger.info(f"Regime sizing: {reduced_pct:.1f}% reduced, {skipped_pct:.1f}% skipped")
    
    ACTIVE_POSITION_COL = "position_regime_adj"
else:
    df_wf["position_base"] = df_wf[ACTIVE_SIGNAL_COL] * POSITION_SIZE
    ACTIVE_POSITION_COL = "position_base"
    logger.info("Regime-aware sizing: DISABLED")


# ------------------------------------------------------------
# 11. Save Backtest Results
# ------------------------------------------------------------

BACKTEST_FILE = ROOT / "artifacts" / "walkforward_backtest.parquet"
df_wf.to_parquet(BACKTEST_FILE)

PERF_FILE = ROOT / "artifacts" / "walkforward_performance.parquet"
df_performance.to_parquet(PERF_FILE)

logger.info(f"Saved backtest to {BACKTEST_FILE}")
logger.info(f"Saved performance comparison to {PERF_FILE}")
# ------------------------------------------------------------
# MATCHING DEBUG
# ------------------------------------------------------------
# Check if signals are identical
print(f"Signals identical? {(df_wf['signal_direct'] == df_wf['signal_thresh']).all()}")
print(f"Matching signals: {(df_wf['signal_direct'] == df_wf['signal_thresh']).sum()} / {len(df_wf)}")
# Verify they're now different
print(f"Direct signals: {df_wf['signal_direct'].value_counts().to_dict()}")
print(f"Threshold signals: {df_wf['signal_thresh'].value_counts().to_dict()}")
# ------------------------------------------------------------
# SPECIFIC STRATEGY DEBUG
# ------------------------------------------------------------
# Signal Values, Exec Signals, Positions // Equity Curve // Drawdown // Daily PnL

print("=== DIRECT STRATEGY PROFILE ===")
print(f"Signal values: {df_wf['signal_direct'].value_counts().to_dict()}")
print(f"Exec signal values: {df_wf['direct_exec'].value_counts().to_dict()}")
print(f"Position values: {df_wf['direct_position'].value_counts().to_dict()}")

print(f"\nEquity curve:")
print(f"  First 5: {df_wf['direct_equity'].head().tolist()}")
print(f"  Last 5: {df_wf['direct_equity'].tail().tolist()}")
print(f"  Min: {df_wf['direct_equity'].min()}, Max: {df_wf['direct_equity'].max()}")

print(f"\nDrawdown:")
print(f"  First 5: {df_wf['direct_drawdown'].head().tolist()}")
print(f"  Last 5: {df_wf['direct_drawdown'].tail().tolist()}")
print(f"  Min: {df_wf['direct_drawdown'].min()}, Max: {df_wf['direct_drawdown'].max()}")
print(f"  Unique values count: {df_wf['direct_drawdown'].nunique()}")

daily_pnl = df_wf['direct_pnl_net'].resample('1D').sum()
print(f"\nDaily PnL:")
print(f"  Count: {len(daily_pnl)}")
print(f"  Non-zero days: {(daily_pnl != 0).sum()}")
print(f"  NaN days: {daily_pnl.isna().sum()}")

print("\n=== THRESHOLD STRATEGY PROFILE ===")
print(f"Signal values: {df_wf['signal_thresh'].value_counts().to_dict()}")
print(f"Exec signal values: {df_wf['threshold_exec'].value_counts().to_dict()}")
print(f"Position values: {df_wf['threshold_position'].value_counts().to_dict()}")

print(f"\nEquity curve:")
print(f"  First 5: {df_wf['threshold_equity'].head().tolist()}")
print(f"  Last 5: {df_wf['threshold_equity'].tail().tolist()}")
print(f"  Min: {df_wf['threshold_equity'].min()}, Max: {df_wf['threshold_equity'].max()}")

print(f"\nDrawdown")
print(f"  First 5: {df_wf['threshold_drawdown'].head().tolist()}")
print(f"  Last 5: {df_wf['threshold_drawdown'].tail().tolist()}")
print(f"  Min: {df_wf['threshold_drawdown'].min()}, Max: {df_wf['threshold_drawdown'].max()}")
print(f"  Unique values count: {df_wf['threshold_drawdown'].nunique()}")

daily_pnl = df_wf['threshold_pnl_net'].resample('1D').sum()
print(f"\nDaily PnL:")
print(f"  Count: {len(daily_pnl)}")
print(f"  Non-zero days: {(daily_pnl != 0).sum()}")
print(f"  NaN days: {daily_pnl.isna().sum()}")

print("\n=== NET PROBABILITY STRATEGY PROFILE ===")
print(f"Signal values: {df_wf['signal_net'].value_counts().to_dict()}")
print(f"Exec signal values: {df_wf['net_prob_exec'].value_counts().to_dict()}")
print(f"Position values: {df_wf['net_prob_position'].value_counts().to_dict()}")

print(f"\nEquity curve:")
print(f"  First 5: {df_wf['net_prob_equity'].head().tolist()}")
print(f"  Last 5: {df_wf['net_prob_equity'].tail().tolist()}")
print(f"  Min: {df_wf['net_prob_equity'].min()}, Max: {df_wf['net_prob_equity'].max()}")

print(f"\nDrawdown:")
print(f"  First 5: {df_wf['net_prob_drawdown'].head().tolist()}")
print(f"  Last 5: {df_wf['net_prob_drawdown'].tail().tolist()}")
print(f"  Min: {df_wf['net_prob_drawdown'].min()}, Max: {df_wf['net_prob_drawdown'].max()}")
print(f"  Unique values count: {df_wf['net_prob_drawdown'].nunique()}")

daily_pnl = df_wf['net_prob_pnl_net'].resample('1D').sum()
print(f"\nDaily PnL:")
print(f"  Count: {len(daily_pnl)}")
print(f"  Non-zero days: {(daily_pnl != 0).sum()}")
print(f"  NaN days: {daily_pnl.isna().sum()}")

# Check if drawdown is all zeros
print(f"\nIs drawdown all zeros? {(df_wf['direct_drawdown'] == 0).all()}")
print(f"\nIs drawdown all zeros? {(df_wf['threshold_drawdown'] == 0).all()}")
print(f"\nIs drawdown all zeros? {(df_wf['net_prob_drawdown'] == 0).all()}")

# ------------------------------------------------------------
# Done — Section 09
# ------------------------------------------------------------

logger.info("Section 09 complete — Backtesting and execution simulation finished.")
````

````python
# ============================================================
# SECTION 10 — Risk & PnL Analytics
# ============================================================

logger.info("Starting Section 10 — Risk & PnL Analytics")

df_bt = df_wf.copy()
df_bt["timestamp"] = df_bt.index
# === DEBUG: Verify execution delay is working ===
print("Signal vs Exec comparison (first 10 rows):")
print(df_bt[["signal_direct", "direct_exec"]].head(10))

print(f"\nFirst signal: {df_bt['signal_direct'].iloc[0]}")
print(f"First exec: {df_bt['direct_exec'].iloc[0]}")  # Should be 0 or different from signal

# === Check data frequency ===
print(f"Data frequency: {df_bt.index.to_series().diff().mode()[0]}")
print(f"Total rows: {len(df_bt)}")
print(f"Date range: {df_bt.index.min()} to {df_bt.index.max()}")
# === END DEBUG ===

# Strategy definitions (must match Section 09)
strategies = {
    "direct": "signal_direct",
    "threshold": "signal_thresh",
    "net_prob": "signal_net"
}


# ------------------------------------------------------------
# 1. Basic PnL & Risk Metrics (All Strategies)
# ------------------------------------------------------------

logger.info("Computing basic risk metrics for all strategies...")

all_metrics = {}

for strat_name in strategies.keys():
    pnl_col = f"{strat_name}_pnl_net"
    equity_col = f"{strat_name}_equity"
    trade_col = f"{strat_name}_trade"
    
    daily_pnl = df_bt[pnl_col].resample("1D").sum()
    daily_vol = daily_pnl.std() * np.sqrt(252)
    annual_pnl = daily_pnl.sum()
    
    sharpe = (daily_pnl.mean() * np.sqrt(252)) / (daily_pnl.std() + 1e-9)
    downside_returns = daily_pnl[daily_pnl < 0]
    downside_vol = downside_returns.std() * np.sqrt(252) if len(downside_returns) > 0 else 1e-9
    sortino = (daily_pnl.mean() * np.sqrt(252)) / (downside_vol + 1e-9)
    
    max_dd = df_bt[equity_col].cummax().sub(df_bt[equity_col]).max()
    
    all_metrics[strat_name] = {
        "Annual PnL": annual_pnl,
        "Daily Vol": daily_vol,
        "Sharpe": sharpe,
        "Sortino": sortino,
        "Max Drawdown": max_dd,
        "Win Rate (%)": (df_bt[pnl_col] > 0).mean() * 100,
        "Trades": df_bt[trade_col].sum(),
    }

# Add confidence interval to Sharpe
n_days = len(daily_pnl)
sharpe_se = 1 / np.sqrt(n_days)  # Standard error
print(f"Sharpe: {sharpe:.2f} ± {1.96 * sharpe_se:.2f} (95% CI)")
print(f"Sample size: {n_days} days (need 100+ for reliable estimate)")

df_metrics = pd.DataFrame(all_metrics).T
logger.info(f"Risk Metrics Comparison:\n{df_metrics}")

# Display formatted table
print("\n" + "="*70)
print("RISK METRICS COMPARISON — ALL STRATEGIES")
print("="*70)
print(df_metrics.to_string())
print("="*70 + "\n")
# ------------------------------------------------------------
# 2. Monthly & Weekly Performance Breakdown
# ------------------------------------------------------------

logger.info("Computing time-based performance breakdown...")

fig, axes = plt.subplots(2, 1, figsize=(14, 8))

for strat_name in strategies.keys():
    daily_pnl = df_bt[f"{strat_name}_pnl_net"].resample("1D").sum()
    monthly_pnl = daily_pnl.resample("ME").sum()   # Changed M to ME
    weekly_pnl = daily_pnl.resample("W").sum()
    
    axes[0].plot(monthly_pnl, marker="o", label=strat_name.title())
    axes[1].plot(weekly_pnl, marker=".", alpha=0.7, label=strat_name.title())

axes[0].set_title("Monthly PnL — All Strategies")
axes[0].set_ylabel("PnL ($)")
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].set_title("Weekly PnL — All Strategies")
axes[1].set_ylabel("PnL ($)")
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
# ------------------------------------------------------------
# 3. Drawdown Analysis
# ------------------------------------------------------------

logger.info("Computing drawdown curves...")

fig, axes = plt.subplots(len(strategies), 1, figsize=(14, 4 * len(strategies)))

for i, strat_name in enumerate(strategies.keys()):
    equity = df_bt[f"{strat_name}_equity"]
    rolling_max = equity.cummax()
    drawdown = equity - rolling_max
    
    axes[i].fill_between(df_bt.index, drawdown, color="red", alpha=0.3)
    axes[i].set_title(f"Drawdown Curve — {strat_name.title()}")
    axes[i].set_ylabel("Drawdown ($)")
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Combined drawdown comparison
plt.figure(figsize=(14, 5))
for strat_name in strategies.keys():
    equity = df_bt[f"{strat_name}_equity"]
    drawdown = equity - equity.cummax()
    plt.plot(drawdown, label=strat_name.title())
plt.title("Drawdown Comparison — All Strategies")
plt.ylabel("Drawdown ($)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()


# ------------------------------------------------------------
# 4. Regime-Based PnL Attribution
# ------------------------------------------------------------

logger.info("Computing regime-based PnL decomposition...")

# Use Regime_Combo from features (X), not df_labels
df_reg = df_bt.join(X["Regime_Combo"], how="left")

regime_reports = {}

for strat_name in strategies.keys():
    pnl_col = f"{strat_name}_pnl_net"
    
    regime_pnl = df_reg.groupby("Regime_Combo")[pnl_col].sum()
    regime_counts = df_reg.groupby("Regime_Combo")[pnl_col].count()
    
    regime_reports[strat_name] = pd.DataFrame({
        "PnL": regime_pnl,
        "Samples": regime_counts,
        "Avg PnL": regime_pnl / regime_counts
    })

# Plot regime PnL for best strategy (by Sharpe)
best_strat = df_metrics["Sharpe"].idxmax()
logger.info(f"Regime PnL report for {best_strat}:\n{regime_reports[best_strat]}")

fig, axes = plt.subplots(1, len(strategies), figsize=(5 * len(strategies), 5))
for i, strat_name in enumerate(strategies.keys()):
    regime_reports[strat_name]["PnL"].plot(kind="bar", ax=axes[i])
    axes[i].set_title(f"PnL by Regime — {strat_name.title()}")
    axes[i].set_ylabel("PnL ($)")
    axes[i].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# 5. Volatility Targeting (10% annualized)
# ------------------------------------------------------------

logger.info("Applying volatility targeting (target 10% annualized)...")

TARGET_VOL = 0.10

plt.figure(figsize=(14, 6))

for strat_name in strategies.keys():
    pnl_col = f"{strat_name}_pnl_net"
    daily_pnl = df_bt[pnl_col].resample("1D").sum()
    real_vol = daily_pnl.std() * np.sqrt(252)
    scaling_factor = TARGET_VOL / (real_vol + 1e-9)
    
    df_bt[f"{strat_name}_pnl_vol_targeted"] = df_bt[pnl_col] * scaling_factor
    df_bt[f"{strat_name}_equity_vol_targeted"] = df_bt[f"{strat_name}_pnl_vol_targeted"].cumsum()
    
    plt.plot(df_bt[f"{strat_name}_equity_vol_targeted"], label=f"{strat_name.title()} (Vol-Targeted)")

plt.title("Equity Curves — Volatility Targeted (10%)")
plt.ylabel("Equity ($)")
plt.xlabel("Time")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()


# ------------------------------------------------------------
# 6. Distributional Analytics
# ------------------------------------------------------------

logger.info("Computing distributional characteristics...")

import scipy.stats as st

dist_stats = {}

fig, axes = plt.subplots(2, len(strategies), figsize=(5 * len(strategies), 8))

for i, strat_name in enumerate(strategies.keys()):
    pnl_dist = df_bt[f"{strat_name}_pnl_net"].dropna()
    
    dist_stats[strat_name] = {
        "Mean": pnl_dist.mean(),
        "Std Dev": pnl_dist.std(),
        "Skew": pnl_dist.skew(),
        "Kurtosis": pnl_dist.kurtosis(),
    }
    
    # Histogram
    sns.histplot(pnl_dist, bins=50, ax=axes[0, i])
    axes[0, i].set_title(f"PnL Distribution — {strat_name.title()}")
    axes[0, i].set_xlabel("PnL ($)")
    
    # QQ Plot
    st.probplot(pnl_dist, dist="norm", plot=axes[1, i])
    axes[1, i].set_title(f"QQ Plot — {strat_name.title()}")

plt.tight_layout()
plt.show()

df_dist_stats = pd.DataFrame(dist_stats).T
logger.info(f"Distribution Stats:\n{df_dist_stats}")


# ------------------------------------------------------------
# 7. Exposure & Turnover Metrics
# ------------------------------------------------------------

logger.info("Computing exposure and turnover...")

exposure_reports = {}

for strat_name in strategies.keys():
    pos_col = f"{strat_name}_position"
    trade_col = f"{strat_name}_trade"
    
    exposure = df_bt[pos_col].abs().mean()
    long_pct = (df_bt[pos_col] > 0).mean() * 100
    short_pct = (df_bt[pos_col] < 0).mean() * 100
    flat_pct = (df_bt[pos_col] == 0).mean() * 100
    
    n_days = (df_bt.index.max() - df_bt.index.min()).days
    annual_turnover = df_bt[trade_col].sum() / (n_days / 365) if n_days > 0 else 0
    
    exposure_reports[strat_name] = {
        "Avg Exposure": exposure,
        "% Long": long_pct,
        "% Short": short_pct,
        "% Flat": flat_pct,
        "Annual Turnover": annual_turnover,
    }

df_exposure = pd.DataFrame(exposure_reports).T
logger.info(f"Exposure Report:\n{df_exposure}")

print("\n" + "="*70)
print("EXPOSURE & TURNOVER — ALL STRATEGIES")
print("="*70)
print(df_exposure.to_string())
print("="*70 + "\n")
# ============================================================
# BASE MODEL SAMPLE STATISTICS
# ============================================================

# Compute daily PnL for base model
base_daily_pnl = df_bt["direct_pnl_net"].resample("1D").sum()

print("=" * 70)
print("  BASE MODEL — SAMPLE STATISTICS")
print("=" * 70)
print(f"\n{'Metric':<35} {'Value'}")
print("-" * 70)
print(f"{'Total samples (bars)':<35} {len(df_bt)}")
print(f"{'Daily observations':<35} {len(base_daily_pnl)}")
print(f"{'Date range start':<35} {df_bt.index.min()}")
print(f"{'Date range end':<35} {df_bt.index.max()}")
print(f"{'Calendar days':<35} {(df_bt.index.max() - df_bt.index.min()).days}")
print(f"{'Bars per day (approx)':<35} {len(df_bt) / max(len(base_daily_pnl), 1):.1f}")
print(f"{'Active trading bars':<35} {(df_bt['direct_position'] != 0).sum()}")
print(f"{'Trade count':<35} {df_bt['direct_trade'].sum():.0f}")
print("=" * 70)

# ------------------------------------------------------------
# 8. Strategy Ranking Summary
# ------------------------------------------------------------

logger.info("Computing strategy rankings...")

ranking = pd.DataFrame({
    "Sharpe": df_metrics["Sharpe"].rank(ascending=False),
    "Sortino": df_metrics["Sortino"].rank(ascending=False),
    "Max DD": df_metrics["Max Drawdown"].rank(ascending=True),  # Lower is better
    "Win Rate": df_metrics["Win Rate (%)"].rank(ascending=False),
})
ranking["Avg Rank"] = ranking.mean(axis=1)
ranking = ranking.sort_values("Avg Rank")

print("\n" + "="*70)
print("STRATEGY RANKING (1 = Best)")
print("="*70)
print(ranking.to_string())
print("="*70 + "\n")

best_overall = ranking["Avg Rank"].idxmin()
logger.info(f"Best overall strategy: {best_overall}")
# ============================================================
# REGIME-STRATIFIED BACKTEST METRICS
# ============================================================

logger.info("Computing regime-stratified backtest metrics...")

# Join regime data
df_regime_bt = df_bt.join(X["Regime_Combo"], how="left")

def compute_regime_metrics(g):
    """Compute full metrics for a regime slice."""
    pnl = g["direct_pnl_net"] if "direct_pnl_net" in g.columns else g.get("pnl_net", pd.Series([0]))
    
    daily_pnl = pnl.resample("D").sum().dropna()
    
    metrics = {
        "Total PnL": pnl.sum(),
        "Avg PnL": pnl.mean(),
        "Std PnL": pnl.std(),
        "Sharpe": (np.sqrt(252) * daily_pnl.mean() / daily_pnl.std()) if daily_pnl.std() > 0 else 0,
        "Win Rate": (pnl > 0).mean(),
        "Max DD": (pnl.cumsum() - pnl.cumsum().cummax()).min(),
        "N Trades": len(pnl[pnl != 0]),
        "N Bars": len(g),
    }
    return pd.Series(metrics)

# Compute per regime
regime_performance = df_regime_bt.groupby("Regime_Combo").apply(compute_regime_metrics)
regime_performance = regime_performance.sort_values("Sharpe", ascending=False)

print("\n" + "=" * 80)
print("REGIME-STRATIFIED BACKTEST PERFORMANCE")
print("=" * 80)
print(regime_performance.to_string())
print("=" * 80)

# Identify profitable vs losing regimes
profitable_regimes = regime_performance[regime_performance["Total PnL"] > 0].index.tolist()
losing_regimes = regime_performance[regime_performance["Total PnL"] <= 0].index.tolist()

print(f"\n✅ Profitable regimes: {profitable_regimes}")
print(f"❌ Losing regimes: {losing_regimes}")

# Save for later use
REGIME_PERFORMANCE = regime_performance.copy()

# ============================================================
# REGIME-STRATIFIED PERFORMANCE TABLE (DETAILED)
# ============================================================

logger.info("Building detailed regime performance table...")

def build_regime_table(df, regime_col="Regime_Combo", pnl_col="direct_pnl_net"):
    """Build comprehensive regime performance table."""
    
    df_reg = df.join(X[regime_col], how="left") if regime_col not in df.columns else df
    
    results = []
    
    for regime in sorted(df_reg[regime_col].dropna().unique()):
        mask = df_reg[regime_col] == regime
        g = df_reg[mask]
        pnl = g[pnl_col] if pnl_col in g.columns else g["pnl_net"]
        
        # Daily aggregation for Sharpe
        daily = pnl.resample("D").sum().dropna()
        
        # Compute metrics
        total_pnl = pnl.sum()
        n_trades = (pnl != 0).sum()
        win_rate = (pnl > 0).mean() if len(pnl) > 0 else 0
        avg_win = pnl[pnl > 0].mean() if (pnl > 0).any() else 0
        avg_loss = pnl[pnl < 0].mean() if (pnl < 0).any() else 0
        sharpe = (np.sqrt(252) * daily.mean() / daily.std()) if daily.std() > 0 else 0
        sortino = (np.sqrt(252) * daily.mean() / daily[daily < 0].std()) if (daily < 0).any() and daily[daily < 0].std() > 0 else 0
        max_dd = (pnl.cumsum() - pnl.cumsum().cummax()).min()
        
        results.append({
            "Regime": regime,
            "Total PnL": total_pnl,
            "N Trades": n_trades,
            "Win Rate": win_rate,
            "Avg Win": avg_win,
            "Avg Loss": avg_loss,
            "Profit Factor": abs(avg_win / avg_loss) if avg_loss != 0 else np.inf,
            "Sharpe": sharpe,
            "Sortino": sortino,
            "Max DD": max_dd,
            "Bars": len(g),
            "% of Data": len(g) / len(df_reg) * 100,
        })
    
    return pd.DataFrame(results).set_index("Regime").sort_values("Sharpe", ascending=False)

# Build table
regime_table = build_regime_table(df_bt)

print("\n" + "=" * 100)
print("DETAILED REGIME PERFORMANCE TABLE")
print("=" * 100)
print(regime_table.round(3).to_string())
print("=" * 100)

# Visualization
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Sharpe by regime
regime_table["Sharpe"].plot(kind="bar", ax=axes[0], color=["green" if x > 0 else "red" for x in regime_table["Sharpe"]])
axes[0].set_title("Sharpe Ratio by Regime")
axes[0].axhline(0, color="black", linestyle="--", alpha=0.5)

# Total PnL by regime
regime_table["Total PnL"].plot(kind="bar", ax=axes[1], color=["green" if x > 0 else "red" for x in regime_table["Total PnL"]])
axes[1].set_title("Total PnL by Regime")
axes[1].axhline(0, color="black", linestyle="--", alpha=0.5)

# Win Rate by regime
regime_table["Win Rate"].plot(kind="bar", ax=axes[2], color="steelblue")
axes[2].set_title("Win Rate by Regime")
axes[2].axhline(0.5, color="red", linestyle="--", alpha=0.5, label="50%")

plt.tight_layout()
plt.savefig(VISUAL_SAVE_DIR / "regime_performance.png", dpi=150)
plt.show()

# ------------------------------------------------------------
# Save Section 10 Artifacts
# ------------------------------------------------------------

RISK_FILE = ROOT / "artifacts" / "risk_metrics_all.parquet"
df_metrics.to_parquet(RISK_FILE)

EXPOSURE_FILE = ROOT / "artifacts" / "exposure_report.parquet"
df_exposure.to_parquet(EXPOSURE_FILE)

logger.info(f"Saved risk analytics to {RISK_FILE}")
logger.info(f"Saved exposure report to {EXPOSURE_FILE}")

logger.info("Section 10 complete — Risk and PnL analytics computed for all strategies.")
````

````python
# ============================================================
# SECTION 13 — Meta-Modeling & Performance Optimization
# ============================================================

logger.info("Starting Section 13 — Meta-Modeling & Performance Optimization")


# ------------------------------------------------------------
# 1. Build Meta-Model Training Dataset
# ------------------------------------------------------------

logger.info("Constructing meta-model dataset...")

# Start from walkforward predictions (true out-of-sample)
df_meta = df_wf_preds.copy()
# Print column names for debugging
print(df_wf_preds.columns.tolist())
# Add key model diagnostics as features
# Overall confidence = max probability across all classes
prob_cols = [col for col in df_wf_preds.columns if col.startswith("prob_class_")]
df_meta["confidence"] = df_meta[prob_cols].max(axis=1)
# Proper 3-class entropy
probs = df_meta[prob_cols].values
df_meta["entropy"] = -np.sum(probs * np.log(probs + 1e-9), axis=1)
# Uncertainty = 1 - confidence (simpler than entropy)
df_meta["uncertainty"] = 1 - df_meta["confidence"]
# ------------------------------------------------------------------
# Add regime labels
df_meta = df_meta.join(X["Regime_Combo"], how="left")

# Add forward return direction
df_meta = df_meta.join(df_labels["fwd_dir"], how="left")

# Add volatility-adjusted return target
vol_adj_cols = ["vol_norm_ret", "zscore_ret", "risk_adj_dir", "svr_target", "dd_adj_ret"]
df_meta = df_meta.join(df_labels[vol_adj_cols], how="left")

# Add feature stability ranking (from Section 07)
if "df_stability" in globals():
    df_meta["stability_rank"] = df_stability["stability"].rank().reindex(df_meta.index, fill_value=0)
else:
    logger.warning("df_stability not found; setting stability_rank=0. Run Section 07 to compute.")
    df_meta["stability_rank"] = 0

# Add rolling volatility of predictions
# Use the bullish class probability, or last probability column
bullish_col = "prob_class_1" if "prob_class_1" in df_meta.columns else prob_cols[-1]
meta_window = get_rolling_window(len(df_meta), target_pct=0.10, min_window=10)
df_meta["pred_vol"] = df_meta[bullish_col].rolling(meta_window).std()

# Drop rows with NaN
df_meta.dropna(inplace=True)

logger.info(f"Meta-model dataset shape: {df_meta.shape}")
# ------------------------------------------------------------
# 2. Define Meta-Model Target
# ------------------------------------------------------------
# Hedge funds often use classification meta-model:
# 1 = prediction was correct
# 0 = prediction was wrong

df_meta["meta_target"] = ((df_meta["pred"] != 0) & (df_meta["true"] == df_meta["pred"])).astype(int)

logger.info("Meta-target (updated accuracy filter) generated.")


# ------------------------------------------------------------
# 3. Select Meta-Model Features
# ------------------------------------------------------------

meta_features = [
    *prob_cols,        # Unpacks whatever prob columns exist
    "confidence",      # max probability
    "entropy",         # prediction uncertainty
    "Regime_Combo",    # from X
    "fwd_dir",         # from df_labels
    *vol_adj_cols,         # from df_labels
    "stability_rank",  # from df_stability
    "pred_vol",        # rolling std of prob_class_1
]
X_meta = df_meta[meta_features]
y_meta = df_meta["meta_target"]
# ------------------------------------------------------------
# 4. Train Meta-Model (Logistic Regression + Random Forest)
# ------------------------------------------------------------
logger.info("Training logistic regression meta-model...")

logit_model = LogisticRegression(max_iter=500)
logit_model.fit(X_meta, y_meta)

logger.info("Training random forest meta-model...")

rf_meta = RandomForestClassifier(
    n_estimators=N_ESTIMATORS,
    max_depth=MAX_DEPTH,
    min_samples_leaf=MIN_SAMPLES_LEAF,
    random_state=SEED,
    n_jobs=-1
)
rf_meta.fit(X_meta, y_meta)
# ------------------------------------------------------------
# 5. Evaluate Meta-Model Performance
# ------------------------------------------------------------
logger.info("Evaluating meta-model performance...")

logit_preds = logit_model.predict_proba(X_meta)[:, 1]
rf_preds = rf_meta.predict_proba(X_meta)[:, 1]

logit_auc = roc_auc_score(y_meta, logit_preds)
rf_auc = roc_auc_score(y_meta, rf_preds)

logger.info(f"Logit Meta-Model AUC: {logit_auc:.4f}")
logger.info(f"RF Meta-Model AUC: {rf_auc:.4f}")
# ------------------------------------------------------------
# 6. Use Meta-Model to Filter Primary Signals
# ------------------------------------------------------------

logger.info("Generating meta-filtered trading signals...")

df_meta["meta_prob"] = rf_preds  # using RF meta-model by default
df_meta["meta_signal"] = df_meta["pred"]  # copy base signal

# Only keep signals where meta confidence is high
META_THRESHOLD = df_meta["meta_prob"].quantile(0.3)  # Keep top 70% of confident signals
df_meta.loc[df_meta["meta_prob"] < META_THRESHOLD, "meta_signal"] = 0

logger.info("Meta-filter applied to signals.")
# === ADD DIAGNOSTIC PRINTS HERE ===
print(f"Meta threshold: {META_THRESHOLD}")
print(f"Meta prob range: {df_meta['meta_prob'].min():.3f} to {df_meta['meta_prob'].max():.3f}")
print(f"Signals passing filter: {(df_meta['meta_prob'] >= META_THRESHOLD).sum()} / {len(df_meta)}")
print(f"% signals kept: {(df_meta['meta_prob'] >= META_THRESHOLD).mean() * 100:.1f}%")
print(f"Non-zero meta_signals: {(df_meta['meta_signal'] != 0).sum()}")
# ------------------------------------------------------------
# 2 Table Debug: Check Meta-Features & Columns
# ============================================================
# Table 1: Column Overview
print("=" * 80)
print("  COLUMN OVERVIEW")
print("=" * 80)
print(f"\n{'DataFrame':<20} {'Column Count':<15} {'Columns'}")
print("-" * 80)
print(f"{'meta_features':<20} {len(meta_features):<15} {meta_features}")
print(f"{'X_meta':<20} {len(X_meta.columns):<15} {X_meta.columns.tolist()}")
print(f"{'df_meta':<20} {len(df_meta.columns):<15} {df_meta.columns.tolist()[:10]}...")
print("=" * 80)

# Table 2: Sample Statistics
meta_daily_pnl = df_meta["meta_signal"].resample("D").sum()

print("\n" + "=" * 80)
print("  SAMPLE STATISTICS")
print("=" * 80)
print(f"\n{'Metric':<30} {'Value'}")
print("-" * 80)
print(f"{'Total samples (bars)':<30} {len(df_meta)}")
print(f"{'Daily observations':<30} {len(meta_daily_pnl)}")
print(f"{'Date range start':<30} {df_meta.index.min()}")
print(f"{'Date range end':<30} {df_meta.index.max()}")
print(f"{'Trading days':<30} {(df_meta.index.max() - df_meta.index.min()).days}")
print(f"{'Bars per day (approx)':<30} {len(df_meta) / max(len(meta_daily_pnl), 1):.1f}")
print("=" * 80)

# ============================================================

# ------------------------------------------------------------
# 7. Meta-Filtered PnL Computation
# ------------------------------------------------------------
logger.info("Computing meta-filtered PnL...")

df_meta_bt = df_meta.copy()
df_meta_bt = df_meta_bt.join(df_prices["returns"], how="left")
df_meta_bt["returns"].fillna(0, inplace=True)

# Execution delay
df_meta_bt["exec_signal"] = df_meta_bt["meta_signal"].shift(1).fillna(0)

# PnL
df_meta_bt["price"] = df_prices["Close"].reindex(df_meta_bt.index).ffill()
df_meta_bt["pnl"] = df_meta_bt["exec_signal"] * df_meta_bt["returns"] * df_meta_bt["price"] * POINT_VALUE

# Costs
df_meta_bt["trade"] = df_meta_bt["exec_signal"].diff().fillna(0).abs()
df_meta_bt["cost"] = df_meta_bt["trade"] * (SLIPPAGE + COMMISSION)
df_meta_bt["pnl_net"] = df_meta_bt["pnl"] - df_meta_bt["cost"]

df_meta_bt["equity"] = df_meta_bt["pnl_net"].cumsum()
logger.info("Meta-filtered PnL computed.")
# ------------------------------------------------------------
# 8. Compare Meta-Filtered Equity Curve to Primary Equity Curve
# ------------------------------------------------------------
fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(df_bt["direct_equity"], label="Direct", alpha=0.6)
ax.plot(df_bt["threshold_equity"], label="Threshold", alpha=0.6)
ax.plot(df_bt["net_prob_equity"], label="Net Prob", alpha=0.6)
ax.plot(df_meta_bt["equity"], label="Meta-Filtered", linewidth=2, color="black")
ax.legend()
ax.set_title("Strategy Comparison: Original vs Meta-Filtered")
plt.show()
# ------------------------------------------------------------
# 9. Compute Performance Improvement
# ------------------------------------------------------------
# Sharpe Ratio — Raw vs Bar-Level (Raw & Annualized)
# --- Raw SHARPE ---
meta_daily_pnl = df_meta_bt["pnl_net"].resample("1D").sum()
meta_sharpe_daily_raw = meta_daily_pnl.mean() / (meta_daily_pnl.std() + 1e-9)
meta_sharpe_daily_annual = meta_sharpe_daily_raw * np.sqrt(252)

# --- BAR-LEVEL SHARPE ---
bars_per_day = len(df_meta_bt) / max(len(meta_daily_pnl), 1)
bars_per_year = 252 * bars_per_day

meta_sharpe_bar_raw = df_meta_bt["pnl_net"].mean() / (df_meta_bt["pnl_net"].std() + 1e-9)
meta_sharpe_bar_annual = meta_sharpe_bar_raw * np.sqrt(bars_per_year)

# --- MAX DRAWDOWN ---
meta_max_dd = df_meta_bt["equity"].cummax().sub(df_meta_bt["equity"]).max()

# --- PRINT TABLE ---
print("=" * 60)
print("  META-FILTERED MODEL — SHARPE RATIOS")
print("=" * 60)
print(f"{'Metric':<30} {'Raw':<15} {'Annualized'}")
print("-" * 60)
print(f"{'Sharpe (Daily)':<30} {meta_sharpe_daily_raw:<15.4f} {meta_sharpe_daily_annual:.4f}")
print(f"{'Sharpe (Bar-Level)':<30} {meta_sharpe_bar_raw:<15.4f} {meta_sharpe_bar_annual:.4f}")
print("-" * 60)
print(f"{'Bars per day':<30} {bars_per_day:.1f}")
print(f"{'Daily observations':<30} {len(meta_daily_pnl)}")
print(f"{'Total bars':<30} {len(df_meta_bt)}")
print(f"{'Max Drawdown':<30} ${meta_max_dd:,.2f}")
print("=" * 60)

# ============================================================
# Table 3: Sample Statistics (Base vs Meta)
# ============================================================

# Base model daily PnL (pick one strategy - using "direct")
base_daily_pnl = df_bt["direct_pnl_net"].resample("1D").sum()

print("\n" + "=" * 80)
print("  SAMPLE STATISTICS — BASE vs META-FILTERED")
print("=" * 80)
print(f"\n{'Metric':<35} {'Base Model':<20} {'Meta-Filtered'}")
print("-" * 80)
print(f"{'Total samples (bars)':<35} {len(df_bt):<20} {len(df_meta_bt)}")
print(f"{'Daily observations':<35} {len(base_daily_pnl):<20} {len(meta_daily_pnl)}")
print(f"{'Date range start':<35} {str(df_bt.index.min())[:19]:<20} {str(df_meta_bt.index.min())[:19]}")
print(f"{'Date range end':<35} {str(df_bt.index.max())[:19]:<20} {str(df_meta_bt.index.max())[:19]}")
print(f"{'Bars per day (approx)':<35} {len(df_bt)/max(len(base_daily_pnl),1):<20.1f} {len(df_meta_bt)/max(len(meta_daily_pnl),1):.1f}")
print(f"{'Active trading bars':<35} {(df_bt['direct_position'] != 0).sum():<20} {(df_meta_bt['exec_signal'] != 0).sum()}")
print("=" * 80)
# ------------------------------------------------------------
# 10. Backtest Meta-Filtered Strategy
# ------------------------------------------------------------
logger.info("Backtesting meta-filtered strategy...")

# Create backtest DataFrame from df_meta
df_meta_bt = df_meta.copy()

# Get forward returns (assuming you have this in df_labels or need to calculate)
df_meta_bt["fwd_dir"] = df_labels["fwd_dir"].reindex(df_meta_bt.index)

# Calculate PnL for meta-filtered signals
df_meta_bt["pnl"] = df_meta_bt["meta_signal"] * df_meta_bt["fwd_dir"]

# Apply transaction costs (adjust as needed)
COST_BPS = 5 / 10000  # 5 bps round-trip
df_meta_bt["signal_change"] = df_meta_bt["meta_signal"].diff().abs()
df_meta_bt["cost"] = df_meta_bt["signal_change"] * COST_BPS
df_meta_bt["pnl_net"] = df_meta_bt["pnl"] - df_meta_bt["cost"]

# === ADD DIAGNOSTIC PRINTS HERE ===
print(f"Non-zero meta_signals: {(df_meta_bt['meta_signal'] != 0).sum()}")
print(f"Signal distribution:\n{df_meta_bt['meta_signal'].value_counts()}")
print(f"\nfwd_dir NaN count: {df_meta_bt['fwd_dir'].isna().sum()}")
print(f"fwd_dir zeros: {(df_meta_bt['fwd_dir'] == 0).sum()}")
print(f"fwd_dir range: {df_meta_bt['fwd_dir'].min():.6f} to {df_meta_bt['fwd_dir'].max():.6f}")
print(f"\npnl sum: {df_meta_bt['pnl'].sum():.6f}")
print(f"pnl_net sum: {df_meta_bt['pnl_net'].sum():.6f}")
# Calculate daily PnL for Sharpe
meta_daily_pnl = df_meta_bt["pnl_net"].resample("D").sum()

# Meta-filtered Sharpe ratio (annualized)
meta_sharpe = meta_daily_pnl.mean() / meta_daily_pnl.std() * np.sqrt(252)

# Meta-filtered max drawdown
meta_cumsum = df_meta_bt["pnl_net"].cumsum()
meta_running_max = meta_cumsum.cummax()
meta_drawdown = meta_cumsum - meta_running_max
meta_max_dd = meta_drawdown.min()

logger.info(f"Meta-filtered Sharpe: {meta_sharpe:.3f}")
logger.info(f"Meta-filtered Max DD: {meta_max_dd:.4f}")
# ------------------------------------------------------------
performance_comparison = pd.DataFrame({
    "Metric": ["Sharpe", "Max Drawdown", "Total PnL"],
    "Original": [sharpe, max_dd, df_bt["direct_pnl_net"].sum()],
    "Meta-Filtered": [meta_sharpe, meta_max_dd, df_meta_bt["pnl_net"].sum()],
})
# Add improvement with correct sign for each metric
performance_comparison["Improvement"] = [
    meta_sharpe - sharpe,                           # Higher is better
    max_dd - meta_max_dd,                           # Lower is better (flip sign)
    df_meta_bt["pnl_net"].sum() - df_bt["direct_pnl_net"].sum()  # Higher is better
]
logger.info(f"Performance comparison:\n{performance_comparison}")
# ------------------------------------------------------------
# 11. Save Meta-Model Artifacts
# ------------------------------------------------------------

META_DIR = ROOT / "artifacts" / "meta"
META_DIR.mkdir(exist_ok=True)

performance_comparison.to_json(META_DIR / "meta_performance.json")
df_meta_bt.to_parquet(META_DIR / "meta_backtest.parquet")

logger.info("Saved meta-model artifacts.")

logger.info("Section 13 complete — Meta-modeling and performance optimization finished.")
````

````python
# ============================================================
# SECTION 14 — Ensemble Models & Stacking (Dynamic Classes)
# ============================================================

logger.info("Starting Section 14 — Ensemble Models & Stacking")

# -----------------------------------------------------------------------------
# 14.1 — Dynamic feature extraction utility
# -----------------------------------------------------------------------------

def get_feature_cols(df, exclude_patterns=None, min_valid_pct=0.5):
    """Dynamically extract feature columns, filtering NaN-heavy columns."""
    if exclude_patterns is None:
        exclude_patterns = ['target', 'true', 'pred', 'signal', 'pnl', 'position',
                           'exec', 'cost', 'trade', 'returns', 'price', 'Open',
                           'High', 'Low', 'Close', 'Volume', 'prob_class']
    
    cols = df.columns.tolist()
    feature_cols = [col for col in cols 
                    if not any(p.lower() in col.lower() for p in exclude_patterns)]
    
    n_rows = len(df)
    return [col for col in feature_cols if df[col].notna().sum() / n_rows >= min_valid_pct]

# -----------------------------------------------------------------------------
# 14.2 — Prepare base features and detect classes dynamically
# -----------------------------------------------------------------------------

wf_indices = df_wf.index
X_subset = X.loc[wf_indices]
feature_cols_dynamic = get_feature_cols(X_subset, min_valid_pct=0.8)

X_base = X_subset[feature_cols_dynamic].copy()
y_base_raw = df_wf["true"].copy()

logger.info(f"Dynamic features: {len(feature_cols_dynamic)} of {len(X.columns)} columns")
logger.info(f"Base feature matrix for ensemble: {X_base.shape}")

# Get original class labels
ensemble_classes = sorted(y_base_raw.unique())
n_classes = len(ensemble_classes)
logger.info(f"Original classes: {ensemble_classes} (n={n_classes})")

HAS_BULLISH = 1 in ensemble_classes
HAS_BEARISH = -1 in ensemble_classes
logger.info(f"Bullish: {HAS_BULLISH}, Bearish: {HAS_BEARISH}")

# ENCODE labels to 0, 1, 2... to avoid XGBoost class caching issues
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(ensemble_classes)
y_base = label_encoder.transform(y_base_raw)

logger.info(f"Encoded classes: {list(label_encoder.classes_)} -> {list(range(n_classes))}")

# -----------------------------------------------------------------------------
# 14.3 — Create and fit models
# -----------------------------------------------------------------------------

from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

ens_models = {
    "RandomForest": RandomForestClassifier(n_estimators=N_ESTIMATORS, max_depth=MAX_DEPTH, min_samples_leaf=MIN_SAMPLES_LEAF, random_state=42, n_jobs=-1),
    "XGBoost": XGBClassifier(n_estimators=N_ESTIMATORS, max_depth=MAX_DEPTH, random_state=42, eval_metric='mlogloss'),
    "LightGBM": LGBMClassifier(n_estimators=N_ESTIMATORS, max_depth=MAX_DEPTH, random_state=42, verbose=-1),
    "Logistic": LogisticRegression(max_iter=1000, random_state=42),
    "MLP": MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500, random_state=42),
}

for name, model in ens_models.items():
    model.fit(X_base, y_base)

model_names = list(ens_models.keys())
logger.info("Fitted 5 ensemble models")

# -----------------------------------------------------------------------------
# 14.4 — Helper functions
# -----------------------------------------------------------------------------

def get_encoded_class_idx(class_val):
    """Get encoded index for original class value (-1, 0, 1)."""
    if class_val in label_encoder.classes_:
        return list(label_encoder.classes_).index(class_val)
    return None

def get_class_prob(model, X, orig_class_val):
    """Extract probability for an original class value."""
    encoded_idx = get_encoded_class_idx(orig_class_val)
    if encoded_idx is not None and encoded_idx < len(model.classes_):
        return model.predict_proba(X)[:, encoded_idx]
    return np.zeros(len(X))

def decode_predictions(encoded_preds):
    """Convert encoded predictions back to original class labels."""
    return label_encoder.inverse_transform(encoded_preds.astype(int))

# -----------------------------------------------------------------------------
# 14.5 — Generate predictions and probabilities
# -----------------------------------------------------------------------------

ensemble_preds = pd.DataFrame(index=X_base.index)

for name, model in ens_models.items():
    # Decode predictions back to original labels
    encoded_pred = model.predict(X_base)
    ensemble_preds[name] = decode_predictions(encoded_pred)
    
    if HAS_BULLISH:
        ensemble_preds[f"{name}_prob_bull"] = get_class_prob(model, X_base, 1)
    if HAS_BEARISH:
        ensemble_preds[f"{name}_prob_bear"] = get_class_prob(model, X_base, -1)

# Majority vote (on decoded predictions)
ensemble_preds["majority_vote"] = ensemble_preds[model_names].mode(axis=1)[0]

if HAS_BULLISH:
    bull_cols = [f"{name}_prob_bull" for name in model_names]
    ensemble_preds["avg_prob_bull"] = ensemble_preds[bull_cols].mean(axis=1)
if HAS_BEARISH:
    bear_cols = [f"{name}_prob_bear" for name in model_names]
    ensemble_preds["avg_prob_bear"] = ensemble_preds[bear_cols].mean(axis=1)

if HAS_BULLISH and HAS_BEARISH:
    ensemble_preds["conviction_spread"] = ensemble_preds["avg_prob_bull"] - ensemble_preds["avg_prob_bear"]
    logger.info("Computed conviction spread (bull - bear)")

ensemble_preds["true"] = y_base_raw  # Store original labels
logger.info(f"Ensemble predictions shape: {ensemble_preds.shape}")

# -----------------------------------------------------------------------------
# 14.6 — Evaluate performance
# -----------------------------------------------------------------------------

print("\n" + "="*60)
print("INDIVIDUAL MODEL PERFORMANCE")
print("="*60)

for name in model_names:
    acc = accuracy_score(y_base_raw, ensemble_preds[name])
    print(f"{name:15s} Accuracy: {acc:.4f}")

vote_acc = accuracy_score(y_base_raw, ensemble_preds["majority_vote"])
print(f"{'Majority Vote':15s} Accuracy: {vote_acc:.4f}")

# -----------------------------------------------------------------------------
# 14.7 — Stacking classifier
# -----------------------------------------------------------------------------

stacking_clf = StackingClassifier(
    estimators=[
        ("rf", RandomForestClassifier(n_estimators=N_ESTIMATORS//2, max_depth=MAX_DEPTH, min_samples_leaf=MIN_SAMPLES_LEAF, random_state=42, n_jobs=-1)),
        ("xgb", XGBClassifier(n_estimators=N_ESTIMATORS//2, max_depth=MAX_DEPTH-1, random_state=42, eval_metric='mlogloss')),
        ("lgb", LGBMClassifier(n_estimators=N_ESTIMATORS//2, max_depth=MAX_DEPTH-1, random_state=42, verbose=-1)),
    ],
    final_estimator=LogisticRegression(max_iter=1000, random_state=42),
    cv=CV_FOLDS, n_jobs=-1, passthrough=False
)

stacking_clf.fit(X_base, y_base)  # Uses encoded labels
stacking_preds_encoded = stacking_clf.predict(X_base)
stacking_preds = decode_predictions(stacking_preds_encoded)

stacking_acc = accuracy_score(y_base_raw, stacking_preds)
print(f"\n{'Stacking':15s} Accuracy: {stacking_acc:.4f}")

if HAS_BULLISH:
    ensemble_preds["stacking_prob_bull"] = get_class_prob(stacking_clf, X_base, 1)
if HAS_BEARISH:
    ensemble_preds["stacking_prob_bear"] = get_class_prob(stacking_clf, X_base, -1)

ensemble_preds["stacking_pred"] = stacking_preds

# -----------------------------------------------------------------------------
# 14.8 — Confusion matrix
# -----------------------------------------------------------------------------

print("\n" + "="*60)
print("STACKING CLASSIFIER — CONFUSION MATRIX")
print("="*60)

cm = confusion_matrix(y_base_raw, stacking_preds, labels=ensemble_classes)
print(f"\nClasses: {ensemble_classes}")
print(cm)
print("\nClassification Report:")
print(classification_report(y_base_raw, stacking_preds, labels=ensemble_classes, zero_division=0))

# -----------------------------------------------------------------------------
# 14.9 — Conviction analysis
# -----------------------------------------------------------------------------

if HAS_BULLISH and HAS_BEARISH:
    print("\n" + "="*60)
    print("CONVICTION SPREAD ANALYSIS")
    print("="*60)
    
    spread = ensemble_preds["conviction_spread"]
    print(f"Mean: {spread.mean():.4f}, Std: {spread.std():.4f}")
    print(f"Min / Max: {spread.min():.4f} / {spread.max():.4f}")
    
    high_bull = (spread > 0.3).sum()
    high_bear = (spread < -0.3).sum()
    print(f"\nHigh conviction bull (>0.3):  {high_bull} ({100*high_bull/len(spread):.1f}%)")
    print(f"High conviction bear (<-0.3): {high_bear} ({100*high_bear/len(spread):.1f}%)")

# -----------------------------------------------------------------------------
# 14.10 — Visualizations
# -----------------------------------------------------------------------------

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

ax1 = axes[0, 0]
accuracies = [accuracy_score(y_base_raw, ensemble_preds[name]) for name in model_names] + [vote_acc, stacking_acc]
labels = model_names + ["Majority Vote", "Stacking"]
ax1.barh(labels, accuracies, color=plt.cm.viridis(np.linspace(0, 0.8, len(labels))))
ax1.set_xlabel("Accuracy")
ax1.set_title("Ensemble Model Comparison")
ax1.set_xlim(0, 1)
for i, v in enumerate(accuracies):
    ax1.text(v + 0.01, i, f"{v:.3f}", va="center")

ax2 = axes[0, 1]
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", ax=ax2,
            xticklabels=ensemble_classes, yticklabels=ensemble_classes)
ax2.set_xlabel("Predicted")
ax2.set_ylabel("Actual")
ax2.set_title("Stacking Confusion Matrix")

ax3 = axes[1, 0]
if HAS_BULLISH:
    ax3.hist(ensemble_preds["avg_prob_bull"], bins=30, alpha=0.6, label="P(Bull)", color="green")
if HAS_BEARISH:
    ax3.hist(ensemble_preds["avg_prob_bear"], bins=30, alpha=0.6, label="P(Bear)", color="red")
ax3.set_xlabel("Probability")
ax3.set_ylabel("Frequency")
ax3.set_title("Probability Distributions")
ax3.legend()

ax4 = axes[1, 1]
if HAS_BULLISH and HAS_BEARISH:
    ax4.plot(ensemble_preds.index, ensemble_preds["conviction_spread"], alpha=0.7, lw=0.8)
    ax4.axhline(0, color="black", ls="--", lw=1)
    ax4.axhline(0.3, color="green", ls=":", alpha=0.7, label="High bull")
    ax4.axhline(-0.3, color="red", ls=":", alpha=0.7, label="High bear")
    ax4.set_ylabel("Conviction Spread")
    ax4.set_title("Conviction Over Time")
    ax4.legend()
else:
    ax4.text(0.5, 0.5, f"Only {n_classes} classes", ha="center", va="center", transform=ax4.transAxes)

plt.tight_layout()
plt.show()
# ============================================================
# REGIME-WEIGHTED MODEL ENSEMBLE
# ============================================================

if USE_REGIME_WEIGHTED_MODELS:
    logger.info("Computing regime-weighted model ensemble...")
    
    # First, compute each model's performance per regime
    model_regime_scores = {}
    
    for model_name, model in ens_models.items():
        # Get predictions from this model
        pred_col = f"{model_name}_pred"
        if pred_col not in df_wf.columns:
            continue
        
        df_eval = df_wf.join(X["Regime_Combo"], how="left")
        
        regime_scores = {}
        for regime in df_eval["Regime_Combo"].dropna().unique():
            mask = df_eval["Regime_Combo"] == regime
            y_true = df_eval.loc[mask, "true"]
            y_pred = df_eval.loc[mask, pred_col]
            
            if len(y_true) > 10:  # Minimum samples
                acc = (y_true == y_pred).mean()
                regime_scores[regime] = acc
        
        model_regime_scores[model_name] = regime_scores
    
    # Convert to DataFrame
    regime_model_matrix = pd.DataFrame(model_regime_scores)
    
    print("\n" + "=" * 80)
    print("MODEL ACCURACY BY REGIME")
    print("=" * 80)
    print(regime_model_matrix.round(3).to_string())
    
    # Best model per regime
    best_model_per_regime = regime_model_matrix.idxmax(axis=1)
    print("\n🏆 Best model per regime:")
    print(best_model_per_regime.to_string())
    
    # Create regime-weighted ensemble predictions
    def regime_weighted_predict(row, regime_col="Regime_Combo"):
        """Select prediction from best-performing model for this regime."""
        regime = row.get(regime_col)
        
        if pd.isna(regime) or regime not in best_model_per_regime.index:
            # Fallback to stacking prediction
            return row.get("stacking_pred", row.get("ensemble_pred", 0))
        
        best_model = best_model_per_regime[regime]
        pred_col = f"{best_model}_pred"
        
        return row.get(pred_col, row.get("stacking_pred", 0))
    
    # Apply regime-weighted ensemble
    df_ensemble = df_wf.join(X["Regime_Combo"], how="left")
    df_ensemble["regime_weighted_pred"] = df_ensemble.apply(regime_weighted_predict, axis=1)
    
    # Compare performance
    base_acc = (df_ensemble["true"] == df_ensemble["stacking_pred"]).mean()
    regime_acc = (df_ensemble["true"] == df_ensemble["regime_weighted_pred"]).mean()
    
    print(f"\n📊 Performance comparison:")
    print(f"   Stacking ensemble:      {base_acc:.4f}")
    print(f"   Regime-weighted:        {regime_acc:.4f}")
    print(f"   Improvement:            {(regime_acc - base_acc) * 100:+.2f}%")
    
    logger.info(f"Regime-weighted ensemble: {base_acc:.4f} → {regime_acc:.4f}")
    
else:
    logger.info("Regime-weighted models: DISABLED")


# -----------------------------------------------------------------------------
# 14.11 — Save artifacts
# -----------------------------------------------------------------------------

ensemble_artifacts = {
    "models": ens_models,
    "stacking_clf": stacking_clf,
    "label_encoder": label_encoder,
    "original_classes": ensemble_classes,
    "n_classes": n_classes,
    "has_bullish": HAS_BULLISH,
    "has_bearish": HAS_BEARISH,
    "feature_cols": feature_cols_dynamic,
}

df_ensemble = ensemble_preds.copy()

logger.info("Section 14 complete — Ensemble modeling finished.")
````

````python
# ============================================================
# ENSEMBLE BACKTEST — Compute ens_sharpe, ens_dd, wf_stack
# ============================================================

wf_stack = df_ensemble.copy()
wf_stack["signal"] = wf_stack["stacking_pred"]

# Join price data from walkforward
wf_stack = wf_stack.join(df_wf[["price"]], how="left")

# Compute returns (price percent change)
wf_stack["returns"] = wf_stack["price"].pct_change().fillna(0)

# Execution delay (signal from previous bar)
wf_stack["exec_signal"] = wf_stack["signal"].shift(1).fillna(0)

# PnL calculation (matches meta-backtest approach)
wf_stack["pnl_gross"] = wf_stack["exec_signal"] * wf_stack["returns"] * wf_stack["price"] * POINT_VALUE * POSITION_SIZE

# Transaction costs
wf_stack["trade"] = wf_stack["exec_signal"].diff().abs().fillna(0)
wf_stack["cost"] = wf_stack["trade"] * (SLIPPAGE * POINT_VALUE + COMMISSION)
wf_stack["pnl_net"] = wf_stack["pnl_gross"] - wf_stack["cost"]

# Equity curve
wf_stack["equity"] = wf_stack["pnl_net"].cumsum()

# Ensemble metrics
ens_daily_pnl = wf_stack["pnl_net"].resample("1D").sum()
ens_sharpe = (ens_daily_pnl.mean() / (ens_daily_pnl.std() + 1e-9)) * np.sqrt(252)

ens_running_max = wf_stack["equity"].cummax()
ens_dd = (wf_stack["equity"] - ens_running_max).min()

logger.info(f"Ensemble Sharpe: {ens_sharpe:.3f}")
logger.info(f"Ensemble Max DD: ${ens_dd:,.2f}")

````
